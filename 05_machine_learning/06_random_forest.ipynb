```python
# ============================================
# CELL 1 - RANDOM FOREST INTRODUCTION
# ============================================

"""
Random Forest is an ensemble learning algorithm.

Ensemble Learning means combining multiple models
to produce a better prediction.

Random Forest works by creating MANY decision trees
and combining their predictions.

Example:

Tree 1 → PASS
Tree 2 → PASS
Tree 3 → FAIL
Tree 4 → PASS

Final Prediction → PASS (majority voting)

Advantages:
✔ Higher accuracy
✔ Reduces overfitting
✔ Works well for large datasets
"""

print("Random Forest Algorithm")


# ============================================
# CELL 2 - IMPORT LIBRARIES
# ============================================

"""
Libraries used:

NumPy → numerical operations
Pandas → dataset handling
Matplotlib → visualization
Scikit-Learn → Random Forest model
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


# ============================================
# CELL 3 - CREATE SAMPLE DATASET
# ============================================

"""
Example dataset:
Hours studied vs pass/fail
"""

data = {
    "Hours_Studied": [1,2,3,4,5,6,7,8],
    "Pass": [0,0,0,0,1,1,1,1]
}

df = pd.DataFrame(data)

print(df)


# ============================================
# CELL 4 - FEATURES AND TARGET
# ============================================

"""
Feature → Hours Studied
Target → Pass / Fail
"""

X = df[["Hours_Studied"]]
y = df["Pass"]

print("Features:")
print(X)

print("\nTarget:")
print(y)


# ============================================
# CELL 5 - TRAIN TEST SPLIT
# ============================================

"""
Split dataset into training and testing data
"""

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

print("Training samples:", len(X_train))
print("Testing samples:", len(X_test))


# ============================================
# CELL 6 - TRAIN RANDOM FOREST MODEL
# ============================================

"""
Create Random Forest model.

n_estimators = number of trees in the forest
"""

model = RandomForestClassifier(n_estimators=100)

model.fit(X_train, y_train)

print("Random Forest model trained")


# ============================================
# CELL 7 - MAKE PREDICTIONS
# ============================================

"""
Predict classes for test dataset
"""

predictions = model.predict(X_test)

print("Actual values:", list(y_test))
print("Predicted values:", predictions)


# ============================================
# CELL 8 - MODEL ACCURACY
# ============================================

"""
Evaluate model performance
"""

accuracy = accuracy_score(y_test, predictions)

print("Model Accuracy:", accuracy)


# ============================================
# CELL 9 - PREDICT NEW VALUE
# ============================================

"""
Predict if a student studying 5 hours
will pass or fail
"""

new_student = np.array([[5]])

prediction = model.predict(new_student)

if prediction[0] == 1:
    print("Prediction: PASS")
else:
    print("Prediction: FAIL")


# ============================================
# CELL 10 - FEATURE IMPORTANCE
# ============================================

"""
Random Forest can also tell which features
are most important for predictions.
"""

importance = model.feature_importances_

print("Feature Importance:", importance)


# ============================================
# CELL 11 - SUMMARY
# ============================================

"""
Summary:

✔ Random Forest uses multiple decision trees
✔ Combines predictions using majority voting
✔ Reduces overfitting
✔ Often more accurate than a single tree
✔ Widely used in real ML applications
"""

print("Random Forest notebook completed")

"""
What You Learned

This notebook covers:

    Ensemble learning

    Random Forest algorithm

    Multiple decision trees

    Majority voting

    Feature importance
"""
```
