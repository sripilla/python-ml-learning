```python
# ============================================
# CELL 1 - OVERFITTING VS UNDERFITTING INTRODUCTION
# ============================================

"""
When we train Machine Learning models,
sometimes the model performs poorly.

Two common problems are:

1️⃣ Underfitting
2️⃣ Overfitting

These problems affect how well the model
generalizes to new unseen data.
"""

print("Overfitting vs Underfitting")


# ============================================
# CELL 2 - UNDERFITTING
# ============================================

"""
Underfitting occurs when the model is too simple
to learn the patterns in the data.

Characteristics:

✔ Poor performance on training data
✔ Poor performance on testing data

Example:
Trying to fit a straight line to complex data.

Cause:
Model is too simple.
"""

print("Underfitting → Model too simple")


# ============================================
# CELL 3 - OVERFITTING
# ============================================

"""
Overfitting occurs when the model learns
the training data TOO well.

The model memorizes the data instead of
learning general patterns.

Characteristics:

✔ Very high training accuracy
✔ Poor testing accuracy

Cause:
Model is too complex.
"""

print("Overfitting → Model too complex")


# ============================================
# CELL 4 - IMPORT LIBRARIES
# ============================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression


# ============================================
# CELL 5 - CREATE SAMPLE DATA
# ============================================

"""
Create sample dataset to demonstrate fitting
"""

X = np.array([1,2,3,4,5,6]).reshape(-1,1)
y = np.array([1,4,9,16,25,36])  # quadratic relationship

plt.scatter(X, y)
plt.title("Original Data")
plt.show()


# ============================================
# CELL 6 - UNDERFITTED MODEL (LINEAR)
# ============================================

"""
Linear regression cannot capture
quadratic relationship properly
"""

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

plt.scatter(X, y)
plt.plot(X, y_pred)

plt.title("Underfitting Example")
plt.show()


# ============================================
# CELL 7 - BETTER MODEL (POLYNOMIAL)
# ============================================

"""
Polynomial regression captures
non-linear relationships
"""

poly = PolynomialFeatures(degree=2)

X_poly = poly.fit_transform(X)

model2 = LinearRegression()
model2.fit(X_poly, y)

y_pred2 = model2.predict(X_poly)

plt.scatter(X, y)
plt.plot(X, y_pred2)

plt.title("Better Fit")
plt.show()


# ============================================
# CELL 8 - BIAS VS VARIANCE
# ============================================

"""
Underfitting → High Bias

Overfitting → High Variance

Bias:
Error caused by overly simple model.

Variance:
Error caused by model being too sensitive
to training data.
"""

print("Underfitting → High Bias")
print("Overfitting → High Variance")


# ============================================
# CELL 9 - HOW TO REDUCE OVERFITTING
# ============================================

"""
Techniques to reduce overfitting:

✔ Use more training data
✔ Use simpler model
✔ Regularization
✔ Cross-validation
✔ Feature selection
"""

print("Ways to reduce overfitting listed above")


# ============================================
# CELL 10 - SUMMARY
# ============================================

"""
Summary:

✔ Underfitting → model too simple
✔ Overfitting → model too complex
✔ Goal → balance bias and variance
✔ Model should generalize to new data
"""

print("Overfitting vs Underfitting notebook completed")


"""
What You Learned

  Important ML concepts:

Underfitting

    Model too simple

    Cannot capture patterns

Overfitting

    Model too complex

    Memorizes training data

Bias–Variance Tradeoff

    High bias → underfitting

    High variance → overfittin
"""
```
