```python
# ============================================
# CELL 1 - DECISION TREE INTRODUCTION
# ============================================

"""
Decision Tree is a supervised Machine Learning algorithm
used for classification and regression.

It works like a flowchart where:
- Each node represents a decision
- Each branch represents an outcome
- Each leaf node represents a final prediction

Example:

            Hours > 4 ?
           /           \
        No               Yes
     FAIL               PASS

Decision Trees split data based on the most informative feature.

Common metrics used for splitting:
1. Entropy
2. Information Gain
3. Gini Index
"""

print("Decision Tree Algorithm")


# ============================================
# CELL 2 - IMPORT LIBRARIES
# ============================================

"""
Libraries used:

NumPy → numerical operations
Pandas → dataset handling
Matplotlib → visualization
Scikit-Learn → Decision Tree model
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree


# ============================================
# CELL 3 - CREATE SAMPLE DATASET
# ============================================

"""
Example dataset:

Hours studied vs pass/fail
"""

data = {
    "Hours_Studied": [1,2,3,4,5,6,7,8],
    "Pass": [0,0,0,0,1,1,1,1]
}

df = pd.DataFrame(data)

print(df)


# ============================================
# CELL 4 - FEATURES AND TARGET
# ============================================

"""
Feature → Hours Studied
Target → Pass/Fail
"""

X = df[["Hours_Studied"]]
y = df["Pass"]

print("Features:")
print(X)

print("\nTarget:")
print(y)


# ============================================
# CELL 5 - TRAIN TEST SPLIT
# ============================================

"""
Split dataset into training and testing
"""

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

print("Training samples:", len(X_train))
print("Testing samples:", len(X_test))


# ============================================
# CELL 6 - TRAIN DECISION TREE MODEL
# ============================================

"""
Create and train the Decision Tree model
"""

model = DecisionTreeClassifier()

model.fit(X_train, y_train)

print("Decision Tree model trained")


# ============================================
# CELL 7 - MAKE PREDICTIONS
# ============================================

"""
Predict classes for test dataset
"""

predictions = model.predict(X_test)

print("Actual values:", list(y_test))
print("Predicted values:", predictions)


# ============================================
# CELL 8 - MODEL ACCURACY
# ============================================

"""
Evaluate model performance
"""

accuracy = accuracy_score(y_test, predictions)

print("Model Accuracy:", accuracy)


# ============================================
# CELL 9 - PREDICT NEW VALUE
# ============================================

"""
Predict if a student studying 5 hours
will pass or fail
"""

new_student = np.array([[5]])

prediction = model.predict(new_student)

if prediction[0] == 1:
    print("Prediction: PASS")
else:
    print("Prediction: FAIL")


# ============================================
# CELL 10 - VISUALIZE DECISION TREE
# ============================================

"""
Plot the Decision Tree structure
"""

plt.figure(figsize=(8,5))

plot_tree(
    model,
    feature_names=["Hours Studied"],
    class_names=["Fail","Pass"],
    filled=True
)

plt.title("Decision Tree Visualization")
plt.show()


# ============================================
# CELL 11 - SUMMARY
# ============================================

"""
Summary:

✔ Decision Trees split data into branches
✔ Each node represents a decision
✔ Leaves represent final predictions
✔ Easy to interpret and visualize
✔ Used in many real-world applications
"""

print("Decision Tree notebook completed")


"""

What You Learned

This notebook covers:

    Decision Tree algorithm

    Tree-based learning

    Feature splitting

    Model training

    Tree visualization

"""
```
