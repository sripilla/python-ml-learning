```python
# =========================================
# CELL 1 - TITLE AND INTRODUCTION
# =========================================
"""
PROBABILITY BASICS

Probability measures how likely an event is to occur.

Key concepts:
- Experiment
- Sample Space
- Event
- Probability
- Conditional Probability
- Independence

Probability values range between 0 and 1.
"""


# =========================================
# CELL 2 - IMPORT LIBRARIES
# =========================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# =========================================
# CELL 3 - BASIC PROBABILITY EXAMPLE
# =========================================
"""
Example: Probability of getting heads when flipping a coin
"""

sample_space = ["H", "T"]

prob_heads = 1 / len(sample_space)

print("Probability of Heads:", prob_heads)


# =========================================
# CELL 4 - DICE PROBABILITY
# =========================================
"""
Probability of getting a number greater than 4
"""

dice = [1,2,3,4,5,6]

favorable = [5,6]

prob = len(favorable) / len(dice)

print("Probability of number > 4:", prob)


# =========================================
# CELL 5 - RANDOM EXPERIMENT (COIN TOSS SIMULATION)
# =========================================
coin = np.random.choice(["H","T"], size=20)

print("Coin toss results:")
print(coin)


# =========================================
# CELL 6 - COUNTING RESULTS
# =========================================
unique, counts = np.unique(coin, return_counts=True)

result = dict(zip(unique, counts))

print("Outcome counts:", result)


# =========================================
# CELL 7 - EMPIRICAL PROBABILITY
# =========================================
prob_heads_empirical = result.get("H",0) / len(coin)

print("Empirical Probability of Heads:", prob_heads_empirical)


# =========================================
# CELL 8 - RANDOM DICE ROLLS
# =========================================
dice_rolls = np.random.randint(1,7, size=50)

dice_rolls


# =========================================
# CELL 9 - DISTRIBUTION OF DICE ROLLS
# =========================================
plt.hist(dice_rolls, bins=6)

plt.title("Distribution of Dice Rolls")
plt.xlabel("Dice Value")
plt.ylabel("Frequency")

plt.show()


# =========================================
# CELL 10 - CONDITIONAL PROBABILITY EXAMPLE
# =========================================
"""
Example:

P(A|B) = Probability of A given B
"""

students = pd.DataFrame({
    "Passed": ["Yes","No","Yes","Yes","No","Yes"],
    "Studied": ["Yes","No","Yes","Yes","No","Yes"]
})

students


# =========================================
# CELL 11 - SIMPLE PROBABILITY ANALYSIS
# =========================================
prob_pass = len(students[students["Passed"]=="Yes"]) / len(students)

print("Probability of Passing:", prob_pass)


# =========================================
# CELL 12 - PRACTICE EXERCISES
# =========================================
"""
TRY THESE EXERCISES

1. Simulate 100 coin tosses and calculate probability of heads
2. Simulate 100 dice rolls and plot histogram
3. Compute probability of getting an even number in dice
4. Calculate empirical probability from simulation
"""
```
