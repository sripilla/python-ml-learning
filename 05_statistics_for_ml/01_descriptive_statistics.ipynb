```python
# =========================================
# CELL 1 - TITLE AND INTRODUCTION
# =========================================
"""
DESCRIPTIVE STATISTICS

Descriptive statistics help summarize and understand datasets.

Key measures include:
- Mean
- Median
- Mode
- Variance
- Standard Deviation

These metrics help understand:
• central tendency
• data spread
• distribution behavior
"""


# =========================================
# CELL 2 - IMPORT LIBRARIES
# =========================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# =========================================
# CELL 3 - CREATE SAMPLE DATA
# =========================================
data = [12, 15, 14, 10, 18, 20, 22, 19, 17, 16]

data


# =========================================
# CELL 4 - MEAN
# =========================================
mean_value = np.mean(data)

print("Mean:", mean_value)


# =========================================
# CELL 5 - MEDIAN
# =========================================
median_value = np.median(data)

print("Median:", median_value)


# =========================================
# CELL 6 - MODE
# =========================================
from scipy import stats

mode_value = stats.mode(data)

print("Mode:", mode_value)


# =========================================
# CELL 7 - VARIANCE
# =========================================
variance_value = np.var(data)

print("Variance:", variance_value)


# =========================================
# CELL 8 - STANDARD DEVIATION
# =========================================
std_value = np.std(data)

print("Standard Deviation:", std_value)


# =========================================
# CELL 9 - DATAFRAME EXAMPLE
# =========================================
df = pd.DataFrame({
    "scores": data
})

df


# =========================================
# CELL 10 - SUMMARY STATISTICS
# =========================================
df.describe()


# =========================================
# CELL 11 - HISTOGRAM (DATA DISTRIBUTION)
# =========================================
plt.hist(data, bins=5)

plt.title("Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()


# =========================================
# CELL 12 - BOXPLOT
# =========================================
sns.boxplot(x=data)

plt.title("Boxplot of Data")

plt.show()


# =========================================
# CELL 13 - PRACTICE EXERCISES
# =========================================
"""
TRY THESE EXERCISES

1. Create a new dataset of 20 numbers
2. Calculate mean, median, and standard deviation
3. Plot histogram of the dataset
4. Plot a boxplot to detect outliers
"""
```
