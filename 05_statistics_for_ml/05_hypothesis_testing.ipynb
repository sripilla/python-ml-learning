```python
# =========================================
# CELL 1 - TITLE AND INTRODUCTION
# =========================================
"""
HYPOTHESIS TESTING

Hypothesis testing is used to make decisions about a population
based on sample data.

Important concepts:

Null Hypothesis (H0)
Assumes no effect or no difference.

Alternative Hypothesis (H1)
Assumes there is an effect or difference.

p-value
Probability of observing the result assuming H0 is true.

If p-value < 0.05 → reject the null hypothesis.
"""


# =========================================
# CELL 2 - IMPORT LIBRARIES
# =========================================
import numpy as np
import pandas as pd
from scipy import stats


# =========================================
# CELL 3 - SAMPLE DATA
# =========================================
"""
Example: test scores from a class
"""

scores = [72, 75, 78, 70, 69, 74, 77, 80, 71, 73]

scores


# =========================================
# CELL 4 - SAMPLE MEAN
# =========================================
mean_score = np.mean(scores)

print("Sample Mean:", mean_score)


# =========================================
# CELL 5 - POPULATION MEAN ASSUMPTION
# =========================================
"""
Assume the population mean is 70
"""

population_mean = 70


# =========================================
# CELL 6 - ONE SAMPLE T TEST
# =========================================
t_stat, p_value = stats.ttest_1samp(scores, population_mean)

print("T Statistic:", t_stat)
print("P Value:", p_value)


# =========================================
# CELL 7 - INTERPRET RESULT
# =========================================
alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")


# =========================================
# CELL 8 - TWO SAMPLE T TEST
# =========================================
"""
Compare scores of two classes
"""

class_A = [72,75,78,70,69]
class_B = [65,68,70,66,64]

t_stat, p_value = stats.ttest_ind(class_A, class_B)

print("T Statistic:", t_stat)
print("P Value:", p_value)


# =========================================
# CELL 9 - PRACTICE EXERCISES
# =========================================
"""
TRY THESE EXERCISES

1. Generate a dataset of 20 numbers
2. Assume a population mean
3. Perform a one-sample t-test
4. Interpret the p-value
"""
```
