```python
"""
1. What is a Function
2. Defining a Function
3. Calling a Function
4. Function Parameters
5. Return Statement
6. Default Parameters
7. Keyword Arguments
8. Multiple Return Values
9. Practice Exercises
"""

# What is a Function
# A function is a reusable block of code that performs a specific task.
# Functions help make programs more organized and reduce repetition.

def greet():
    print("Hello, welcome to Python")


# Calling a Function
# A function runs only when it is called.

greet()


# Defining a Function
# Functions are defined using the 'def' keyword.

def show_message():
    print("Learning Python functions")

show_message()


# Function Parameters
# Parameters allow passing information to a function.

def greet_user(name):
    print("Hello", name)

greet_user("Likitha")
greet_user("John")


# Function with Multiple Parameters

def add_numbers(a, b):
    result = a + b
    print(result)

add_numbers(5, 10)


# Return Statement
# The return statement sends a value back to the caller.

def multiply(a, b):
    return a * b

result = multiply(4, 5)

print(result)


# Default Parameters
# Default values are used if no argument is provided.

def greet_user(name="Guest"):
    print("Hello", name)

greet_user("Likitha")
greet_user()


# Keyword Arguments
# Arguments can be passed using parameter names.

def student_info(name, age):
    print("Name:", name)
    print("Age:", age)

student_info(age=20, name="Likitha")


# Multiple Return Values
# A function can return multiple values.

def calculate(a, b):
    sum_value = a + b
    product = a * b
    return sum_value, product

s, p = calculate(4, 5)

print(s)
print(p)


# Practice Exercise 1
# Create a function that prints a welcome message.

def welcome():
    print("Welcome to Python learning")

welcome()


# Practice Exercise 2
# Create a function that adds two numbers and returns the result.

def add(a, b):
    return a + b

print(add(10, 20))


# Practice Exercise 3
# Create a function that checks if a number is even.

def is_even(num):
    if num % 2 == 0:
        return True
    else:
        return False

print(is_even(4))
print(is_even(7))


# Practice Exercise 4
# Create a function that prints numbers from 1 to n.

def print_numbers(n):
    for i in range(1, n+1):
        print(i)

print_numbers(5)
```
