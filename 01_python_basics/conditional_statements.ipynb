```python
"""
1. What are Conditional Statements
2. The if Statement
3. The if-else Statement
4. The if-elif-else Statement
5. Nested Conditions
6. Comparison Operators
7. Logical Operators
8. Practice Exercises
"""

# What are Conditional Statements
# Conditional statements allow a program to make decisions.
# They execute certain blocks of code only if a condition is True.

x = 10

if x > 5:
    print("x is greater than 5")


# The if Statement
# The if statement runs a block of code when the condition is True.

age = 18

if age >= 18:
    print("You are eligible to vote")


# The if-else Statement
# if runs when the condition is True.
# else runs when the condition is False.

age = 16

if age >= 18:
    print("You can vote")
else:
    print("You cannot vote yet")


# The if-elif-else Statement
# elif allows checking multiple conditions.

marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")


# Nested Conditions
# A conditional statement inside another conditional statement.

age = 20
citizen = True

if age >= 18:
    if citizen:
        print("Eligible to vote")
    else:
        print("Not a citizen")
else:
    print("Too young to vote")


# Comparison Operators
# These operators compare values.

a = 10
b = 20

print(a == b)   # equal to
print(a != b)   # not equal
print(a > b)    # greater than
print(a < b)    # less than
print(a >= b)   # greater than or equal to
print(a <= b)   # less than or equal to


# Logical Operators
# Logical operators combine multiple conditions.

age = 25
has_id = True

if age >= 18 and has_id:
    print("Entry allowed")


temperature = 35

if temperature > 30 or temperature < 10:
    print("Extreme weather")


is_logged_in = False

if not is_logged_in:
    print("Please log in")


# Practice Exercise 1
# Check if a number is positive.

num = 10

if num > 0:
    print("Positive number")


# Practice Exercise 2
# Check if a number is even or odd.

num = 7

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")


# Practice Exercise 3
# Determine the largest of two numbers.

a = 15
b = 20

if a > b:
    print("a is larger")
else:
    print("b is larger")


# Practice Exercise 4
# Simple login check.

username = "admin"
password = "1234"

if username == "admin" and password == "1234":
    print("Login successful")
else:
    print("Invalid credentials")
```
