```python
"""
1. What is File Handling
2. Opening a File
3. Reading a File
4. Writing to a File
5. Appending to a File
6. Using 'with' Statement
7. Reading Lines
8. Practice Exercises
"""

# What is File Handling
# File handling allows a program to read data from files
# and write data to files stored on the system.


# Opening a File
# The open() function is used to open a file.

file = open("sample.txt", "w")

file.write("Hello Python")

file.close()


# Reading a File
# Mode "r" is used to read a file.

file = open("sample.txt", "r")

content = file.read()

print(content)

file.close()


# Writing to a File
# Mode "w" overwrites existing content.

file = open("example.txt", "w")

file.write("Python file handling example")

file.close()


# Appending to a File
# Mode "a" adds new content without deleting existing content.

file = open("example.txt", "a")

file.write("\nAppending new line")

file.close()


# Using 'with' Statement
# Using 'with' automatically closes the file after use.

with open("sample.txt", "r") as file:
    content = file.read()
    print(content)


# Reading Lines
# readlines() reads all lines of a file into a list.

with open("sample.txt", "r") as file:
    lines = file.readlines()
    print(lines)


# Reading file line by line

with open("sample.txt", "r") as file:
    for line in file:
        print(line)


# Practice Exercise 1
# Write text to a file.

with open("practice.txt", "w") as file:
    file.write("Learning Python file handling")


# Practice Exercise 2
# Read the contents of the file.

with open("practice.txt", "r") as file:
    print(file.read())


# Practice Exercise 3
# Append a new line to the file.

with open("practice.txt", "a") as file:
    file.write("\nThis is an additional line")


# Practice Exercise 4
# Read file line by line.

with open("practice.txt", "r") as file:
    for line in file:
        print(line)
```
