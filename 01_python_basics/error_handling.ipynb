```python
"""
1. What is Error Handling
2. Common Python Errors
3. try and except
4. Handling Specific Exceptions
5. try-except-else
6. finally Block
7. Raising Exceptions
8. Practice Exercises
"""

# What is Error Handling
# Errors can occur during program execution.
# Error handling allows the program to continue running instead of crashing.

print("Program started")


# Common Python Errors
# Some common errors include:
# ZeroDivisionError
# ValueError
# TypeError
# IndexError

# Example of ZeroDivisionError

# print(10 / 0)


# try and except
# The try block contains code that might cause an error.
# The except block handles the error.

try:
    x = 10 / 0
except:
    print("An error occurred")


# Handling Specific Exceptions
# It is better to catch specific exceptions.

try:
    num = int("abc")
except ValueError:
    print("Invalid number")


# Another example

try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")


# try-except-else
# The else block runs if no exception occurs.

try:
    x = 10 / 2
except ZeroDivisionError:
    print("Division error")
else:
    print("Result:", x)


# finally Block
# The finally block always executes whether an error occurs or not.

try:
    x = 10 / 2
except:
    print("Error occurred")
finally:
    print("Execution completed")


# Raising Exceptions
# You can manually raise an exception using raise.

age = -5

if age < 0:
    raise ValueError("Age cannot be negative")


# Practice Exercise 1
# Handle division by zero.

try:
    num = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")


# Practice Exercise 2
# Handle invalid integer conversion.

try:
    value = int("hello")
except ValueError:
    print("Conversion failed")


# Practice Exercise 3
# Use finally block.

try:
    x = 5 / 1
except:
    print("Error")
finally:
    print("Program finished")
```
