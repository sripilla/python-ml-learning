```python
"""
1. What is List Comprehension
2. Basic List Comprehension
3. List Comprehension with Condition
4. List Comprehension with if-else
5. Using List Comprehension with Strings
6. Nested List Comprehension
7. Practice Exercises
"""

# What is List Comprehension
# List comprehension provides a concise way to create lists.
# It replaces loops used for building lists.

numbers = [1, 2, 3, 4, 5]

squares = []

for n in numbers:
    squares.append(n*n)

print(squares)


# Basic List Comprehension
# Syntax: [expression for item in iterable]

numbers = [1, 2, 3, 4, 5]

squares = [n*n for n in numbers]

print(squares)


# Another example

numbers = [1, 2, 3, 4, 5]

double = [n*2 for n in numbers]

print(double)


# List Comprehension with Condition
# You can filter values using an if condition.

numbers = [1, 2, 3, 4, 5, 6]

even_numbers = [n for n in numbers if n % 2 == 0]

print(even_numbers)


# Another example

numbers = [10, 15, 20, 25, 30]

greater_than_20 = [n for n in numbers if n > 20]

print(greater_than_20)


# List Comprehension with if-else

numbers = [1, 2, 3, 4, 5]

result = ["Even" if n % 2 == 0 else "Odd" for n in numbers]

print(result)


# Using List Comprehension with Strings

word = "python"

letters = [letter.upper() for letter in word]

print(letters)


# Nested List Comprehension

matrix = [[1,2,3],[4,5,6],[7,8,9]]

flatten = [num for row in matrix for num in row]

print(flatten)


# Practice Exercise 1
# Create a list of squares from 1 to 10.

squares = [x*x for x in range(1,11)]

print(squares)


# Practice Exercise 2
# Create a list of even numbers from 1 to 20.

even_numbers = [x for x in range(1,21) if x % 2 == 0]

print(even_numbers)


# Practice Exercise 3
# Convert all words to uppercase.

words = ["python","machine","learning"]

upper_words = [w.upper() for w in words]

print(upper_words)
```
