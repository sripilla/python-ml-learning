```python
"""
1. What are Loops
2. The for Loop
3. Using range()
4. Looping Through Lists
5. The while Loop
6. break Statement
7. continue Statement
8. Nested Loops
9. Practice Exercises
"""

# What are Loops
# Loops allow repeating a block of code multiple times.
# They are useful when performing operations on collections of data.

for i in range(5):
    print("Hello")


# The for Loop
# A for loop is used to iterate over a sequence such as a list or range.

for i in range(5):
    print(i)


# Using range()
# range() generates a sequence of numbers.

for i in range(1, 6):
    print(i)


# Looping Through Lists
# for loops can iterate through elements in a list.

fruits = ["apple", "banana", "mango"]

for fruit in fruits:
    print(fruit)


# The while Loop
# A while loop runs as long as the condition is True.

count = 1

while count <= 5:
    print(count)
    count += 1


# break Statement
# break stops the loop immediately.

for i in range(10):
    if i == 5:
        break
    print(i)


# continue Statement
# continue skips the current iteration and moves to the next one.

for i in range(5):
    if i == 2:
        continue
    print(i)


# Nested Loops
# A loop inside another loop.

for i in range(3):
    for j in range(2):
        print(i, j)


# Practice Exercise 1
# Print numbers from 1 to 10.

for i in range(1, 11):
    print(i)


# Practice Exercise 2
# Print all elements of a list.

numbers = [10, 20, 30, 40]

for num in numbers:
    print(num)


# Practice Exercise 3
# Print even numbers from 1 to 10.

for i in range(1, 11):
    if i % 2 == 0:
        print(i)


# Practice Exercise 4
# Print numbers from 5 to 1 using while loop.

num = 5

while num > 0:
    print(num)
    num -= 1
```
