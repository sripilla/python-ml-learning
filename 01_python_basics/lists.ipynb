```python
"""
1. What is a List
2. Creating Lists
3. Accessing List Elements
4. Negative Indexing
5. List Slicing
6. Modifying Lists
7. Adding Elements
8. Removing Elements
9. List Length
10. Looping Through Lists
11. List Methods
12. Practice Exercises
"""

# What is a List
# A list is a collection of multiple values stored in a single variable.
# Lists are ordered, changeable (mutable), and allow duplicate values.

numbers = [10, 20, 30, 40]

print(numbers)
print(type(numbers))


# Creating Lists
# Lists are created using square brackets []

fruits = ["apple", "banana", "mango"]

print(fruits)


# Lists can store mixed data types

mixed_list = [10, "Python", 3.5, True]

print(mixed_list)


# Accessing List Elements
# List elements are accessed using index numbers.
# Indexing starts from 0.

fruits = ["apple", "banana", "mango"]

print(fruits[0])
print(fruits[1])
print(fruits[2])


# Negative Indexing
# Negative indexing allows accessing elements from the end of the list.

print(fruits[-1])
print(fruits[-2])


# List Slicing
# Slicing allows extracting a portion of a list.

numbers = [10, 20, 30, 40, 50]

print(numbers[1:4])
print(numbers[:3])
print(numbers[2:])


# Modifying Lists
# Lists are mutable, meaning their elements can be changed.

numbers = [10, 20, 30]

numbers[1] = 25

print(numbers)


# Adding Elements
# append() adds an element to the end of the list

numbers = [10, 20, 30]

numbers.append(40)

print(numbers)


# insert() adds an element at a specific index

numbers.insert(1, 15)

print(numbers)


# Removing Elements
# remove() removes a specific element

numbers.remove(20)

print(numbers)


# pop() removes element using index

numbers.pop(2)

print(numbers)


# List Length
# len() returns the number of elements in a list

numbers = [10, 20, 30, 40, 50]

print(len(numbers))


# Looping Through Lists
# Lists can be traversed using a for loop.

numbers = [10, 20, 30, 40]

for num in numbers:
    print(num)


# List Methods
# Some common list methods are append(), insert(), remove(), pop(), sort(), reverse().

numbers = [5, 3, 8, 1]

numbers.sort()

print(numbers)


numbers.reverse()

print(numbers)


# Practice Exercise 1
# Create a list of three cities and print them.

cities = ["Delhi", "Mumbai", "Bangalore"]

print(cities)


# Practice Exercise 2
# Add a new number to the list.

numbers = [1, 2, 3]

numbers.append(4)

print(numbers)


# Practice Exercise 3
# Print each fruit in the list using a loop.

fruits = ["apple", "banana", "mango"]

for fruit in fruits:
    print(fruit)


# Practice Exercise 4
# Find the length of a list.

numbers = [10, 20, 30, 40]

print(len(numbers))
```
