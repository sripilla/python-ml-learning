```python
"""
1. What are Data Types
2. Integer (int)
3. Float (float)
4. String (str)
5. Boolean (bool)
6. Checking Data Types
7. Type Conversion
8. Practice Exercises
"""

# Data types tell Python what kind of value a variable is storing.
# Common built-in data types:
# int   -> whole numbers
# float -> decimal numbers
# str   -> text
# bool  -> True or False


# Integer (int)
# Integers are whole numbers without decimals.

age = 20
year = 2026
temperature = -5

print(age)
print(year)
print(temperature)
print(type(age))


# Float (float)
# Floats represent numbers with decimal points.

height = 5.5
price = 99.99
temperature = -2.3

print(height)
print(price)
print(type(price))


# String (str)
# Strings represent text and are written using single or double quotes.

name = "Likitha"
city = "Manipal"

print(name)
print(city)
print(type(name))


# String concatenation
# Strings can be combined using the + operator.

first_name = "Likitha"
last_name = "Sai"

full_name = first_name + " " + last_name

print(full_name)


# Boolean (bool)
# Boolean values represent True or False.

is_student = True
is_logged_in = False

print(is_student)
print(is_logged_in)
print(type(is_student))


# Boolean comparison example

x = 10
y = 20

print(x > y)
print(x < y)


# Checking Data Types
# The type() function is used to check the data type of a variable.

a = 10
b = 3.14
c = "Python"
d = True

print(type(a))
print(type(b))
print(type(c))
print(type(d))


# Type Conversion
# Python allows converting one data type to another.

# Convert integer to float

x = 10
y = float(x)

print(y)
print(type(y))


# Convert float to integer

x = 5.8
y = int(x)

print(y)


# Convert integer to string

age = 20
age_str = str(age)

print(age_str)
print(type(age_str))


# Practice Exercise 1

name = "Likitha"
age = 20
height = 5.5

print(name)
print(age)
print(height)


# Practice Exercise 2
# Check data types of variables

a = 10
b = 3.14
c = "Hello"
d = False

print(type(a))
print(type(b))
print(type(c))
print(type(d))


# Practice Exercise 3
# Convert number to string

num = 100

num_str = str(num)

print(num_str)
print(type(num_str))
```
