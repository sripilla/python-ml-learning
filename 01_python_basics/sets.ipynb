```python
"""
1. What is a Set
2. Creating Sets
3. Duplicate Removal
4. Accessing Set Elements
5. Adding Elements
6. Removing Elements
7. Set Length
8. Looping Through Sets
9. Set Operations
10. Practice Exercises
"""

# What is a Set
# A set is a collection of unique elements.
# Sets do not allow duplicate values.
# Sets are unordered, meaning elements do not have a fixed position.

numbers = {10, 20, 30, 40}

print(numbers)
print(type(numbers))


# Creating Sets
# Sets are created using curly braces {} or the set() function.

fruits = {"apple", "banana", "mango"}

print(fruits)


# Duplicate Removal
# Sets automatically remove duplicate values.

numbers = {1, 2, 2, 3, 4, 4, 5}

print(numbers)


# Accessing Set Elements
# Since sets are unordered, we cannot access elements using indexing.
# But we can loop through the set.

fruits = {"apple", "banana", "mango"}

for fruit in fruits:
    print(fruit)


# Adding Elements
# add() adds a single element to the set.

numbers = {10, 20, 30}

numbers.add(40)

print(numbers)


# Removing Elements
# remove() deletes a specific element from the set.

numbers.remove(20)

print(numbers)


# discard() also removes an element but does not cause an error if the element does not exist.

numbers.discard(100)

print(numbers)


# Set Length
# len() returns the number of elements in a set.

numbers = {10, 20, 30, 40}

print(len(numbers))


# Looping Through Sets

colors = {"red", "green", "blue"}

for color in colors:
    print(color)


# Set Operations
# Sets support mathematical operations like union and intersection.

A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Union combines elements from both sets
print(A.union(B))

# Intersection returns common elements
print(A.intersection(B))

# Difference returns elements present in one set but not the other
print(A.difference(B))


# Practice Exercise 1
# Create a set of three cities.

cities = {"Delhi", "Mumbai", "Bangalore"}

print(cities)


# Practice Exercise 2
# Add a new city to the set.

cities.add("Chennai")

print(cities)


# Practice Exercise 3
# Remove an element from the set.

cities.remove("Delhi")

print(cities)


# Practice Exercise 4
# Perform union operation.

A = {1, 2, 3}
B = {3, 4, 5}

print(A.union(B))
```
