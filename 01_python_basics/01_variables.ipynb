```python

"""
1. What is a Variable
2. Creating Variables
3. Dynamic Typing
4. Multiple Assignment
5. Same Value Assignment
6. Naming Rules
7. Naming Best Practices
8. Updating Variables
9. Swapping Variables
10. Practice Exercises
"""
# A variable is a name used to store a value in memory.
# Think of it like a labeled box that stores data.

age = 20
name = "Likitha"
height = 5.5
is_student = True

print(age)
print(name)
print(height)
print(is_student)


# Python automatically understands the data type.
x = 10
print(type(x))

x = "Hello"
print(type(x))


# Python allows assigning multiple variables at once.
a, b, c = 1, 2, 3

print(a)
print(b)
print(c)

# Assign Same Value to Multiple Variables
x = y = z = 100

print(x)
print(y)
print(z)

# Variables:
# Must start with letter or underscore
# Can contain letters, numbers, underscores
age = 20
user_name = "Likitha"
height_cm = 165

# Updating Variables
# Variables can change values.

score = 10
score = score + 5
print(score)


# Swapping Variables:

a = 5
b = 10

a, b = b, a

print(a)
print(b)

# Exercise 1
name = "Likitha"
age = 20

print(name, age)



# Exercise 2
length = 5
width = 3

area = length * width
print(area)



# Exercise 3
x = 10
y = 20

temp = x
x = y
y = temp

print(x, y)
```
