```python
"""
1. What is a Dictionary
2. Creating Dictionaries
3. Accessing Dictionary Values
4. Adding Elements
5. Modifying Elements
6. Removing Elements
7. Dictionary Keys and Values
8. Dictionary Length
9. Looping Through Dictionaries
10. Practice Exercises
"""

# What is a Dictionary
# A dictionary is a collection of key-value pairs.
# Each key is unique and is used to access its corresponding value.
# Dictionaries are unordered and mutable (can be changed).

student = {
    "name": "Likitha",
    "age": 20,
    "course": "AI/ML"
}

print(student)
print(type(student))


# Creating Dictionaries
# Dictionaries are created using curly braces {}

person = {
    "name": "John",
    "age": 25,
    "city": "Delhi"
}

print(person)


# Accessing Dictionary Values
# Values can be accessed using their keys.

student = {
    "name": "Likitha",
    "age": 20,
    "course": "AI/ML"
}

print(student["name"])
print(student["age"])


# Adding Elements
# New key-value pairs can be added to a dictionary.

student = {
    "name": "Likitha",
    "age": 20
}

student["course"] = "AI/ML"

print(student)


# Modifying Elements
# Existing values can be updated using their key.

student["age"] = 21

print(student)


# Removing Elements
# pop() removes an element using the key.

student.pop("course")

print(student)


# Dictionary Keys and Values
# keys() returns all keys.
# values() returns all values.

student = {
    "name": "Likitha",
    "age": 20,
    "course": "AI/ML"
}

print(student.keys())
print(student.values())


# Dictionary Length
# len() returns the number of key-value pairs.

print(len(student))


# Looping Through Dictionaries
# Dictionaries can be traversed using a for loop.

student = {
    "name": "Likitha",
    "age": 20,
    "course": "AI/ML"
}

for key in student:
    print(key, student[key])


# Practice Exercise 1
# Create a dictionary for a book.

book = {
    "title": "Python Basics",
    "author": "John Doe",
    "year": 2024
}

print(book)


# Practice Exercise 2
# Access the value of "author".

print(book["author"])


# Practice Exercise 3
# Add a new key "price" to the dictionary.

book["price"] = 500

print(book)
```
