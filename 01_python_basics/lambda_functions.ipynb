```python
"""
1. What is a Lambda Function
2. Basic Lambda Function
3. Lambda with Multiple Arguments
4. Lambda with map()
5. Lambda with filter()
6. Lambda with sorted()
7. Practice Exercises
"""

# What is a Lambda Function
# A lambda function is a small anonymous function.
# It can take any number of arguments but has only one expression.
# Lambda functions are defined using the keyword 'lambda'.

add = lambda a, b: a + b

print(add(5, 10))


# Basic Lambda Function
# Lambda functions can be used to perform simple operations.

square = lambda x: x * x

print(square(4))


# Lambda with Multiple Arguments

multiply = lambda a, b, c: a * b * c

print(multiply(2, 3, 4))


# Lambda with map()
# map() applies a function to every element in an iterable.

numbers = [1, 2, 3, 4, 5]

squares = list(map(lambda x: x*x, numbers))

print(squares)


# Lambda with filter()
# filter() selects elements that satisfy a condition.

numbers = [1, 2, 3, 4, 5, 6]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print(even_numbers)


# Lambda with sorted()
# Lambda can be used as a key function in sorting.

students = [
    ("Alice", 85),
    ("Bob", 75),
    ("Charlie", 90)
]

sorted_students = sorted(students, key=lambda x: x[1])

print(sorted_students)


# Practice Exercise 1
# Create a lambda function that doubles a number.

double = lambda x: x * 2

print(double(6))


# Practice Exercise 2
# Use lambda with map() to cube numbers.

numbers = [1, 2, 3, 4]

cubes = list(map(lambda x: x**3, numbers))

print(cubes)


# Practice Exercise 3
# Use lambda with filter() to get numbers greater than 10.

numbers = [5, 10, 15, 20, 25]

result = list(filter(lambda x: x > 10, numbers))

print(result)
```
