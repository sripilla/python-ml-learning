```python
"""
1. What is a Tuple
2. Creating Tuples
3. Accessing Tuple Elements
4. Negative Indexing
5. Tuple Slicing
6. Tuple Immutability
7. Tuple Packing and Unpacking
8. Tuple Length
9. Looping Through Tuples
10. Practice Exercises
"""

# What is a Tuple
# A tuple is a collection of multiple values stored in a single variable.
# Tuples are ordered and allow duplicate values.
# Unlike lists, tuples are immutable (cannot be changed after creation).

numbers = (10, 20, 30, 40)

print(numbers)
print(type(numbers))


# Creating Tuples
# Tuples are created using parentheses ()

fruits = ("apple", "banana", "mango")

print(fruits)


# A tuple can contain different data types

mixed_tuple = (10, "Python", 3.14, True)

print(mixed_tuple)


# Accessing Tuple Elements
# Tuple elements are accessed using indexing.
# Index starts from 0.

fruits = ("apple", "banana", "mango")

print(fruits[0])
print(fruits[1])
print(fruits[2])


# Negative Indexing
# Negative indexing accesses elements from the end.

print(fruits[-1])
print(fruits[-2])


# Tuple Slicing
# Slicing allows retrieving a portion of a tuple.

numbers = (10, 20, 30, 40, 50)

print(numbers[1:4])
print(numbers[:3])
print(numbers[2:])


# Tuple Immutability
# Tuples cannot be modified after creation.

numbers = (10, 20, 30)

# numbers[1] = 25  # This would cause an error

print(numbers)


# Tuple Packing
# Packing means putting multiple values into a tuple.

person = ("Likitha", 20, "Student")

print(person)


# Tuple Unpacking
# Unpacking means assigning tuple values to variables.

name, age, role = person

print(name)
print(age)
print(role)


# Tuple Length
# len() returns the number of elements in a tuple.

numbers = (10, 20, 30, 40)

print(len(numbers))


# Looping Through Tuples
# Tuples can be iterated using a for loop.

numbers = (10, 20, 30)

for num in numbers:
    print(num)


# Practice Exercise 1
# Create a tuple of three colors and print them.

colors = ("red", "green", "blue")

print(colors)


# Practice Exercise 2
# Access the first element of the tuple.

numbers = (100, 200, 300)

print(numbers[0])


# Practice Exercise 3
# Unpack a tuple into three variables.

person = ("John", 25, "Engineer")

name, age, job = person

print(name)
print(age)
print(job)
```
