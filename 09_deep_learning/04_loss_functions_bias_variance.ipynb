```python
# ==========================================================
# CELL 1 — INTRODUCTION TO LOSS FUNCTIONS
# ==========================================================

# A loss function measures how far the predicted value
# is from the actual value.

# The goal of training a neural network is to minimize
# this loss.

print("Loss functions measure prediction error.")


# ==========================================================
# CELL 2 — MEAN SQUARED ERROR (MSE)
# ==========================================================

# MSE is commonly used for regression problems.

# Formula:
# MSE = (1/n) * Σ(y_true - y_pred)^2

import numpy as np

y_true = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])

mse = np.mean((y_true - y_pred) ** 2)

print("Mean Squared Error:", mse)


# ==========================================================
# CELL 3 — MEAN ABSOLUTE ERROR (MAE)
# ==========================================================

# MAE measures the average absolute difference
# between predicted and true values.

mae = np.mean(np.abs(y_true - y_pred))

print("Mean Absolute Error:", mae)


# ==========================================================
# CELL 4 — CROSS ENTROPY LOSS
# ==========================================================

# Cross entropy is commonly used for classification.

# Example: binary classification

y_true = np.array([1, 0, 1, 1])
y_pred = np.array([0.9, 0.2, 0.8, 0.7])

epsilon = 1e-10

cross_entropy = -np.mean(
    y_true * np.log(y_pred + epsilon) +
    (1 - y_true) * np.log(1 - y_pred + epsilon)
)

print("Binary Cross Entropy:", cross_entropy)


# ==========================================================
# CELL 5 — MODEL COMPLEXITY
# ==========================================================

# When building models we must balance:
# Model simplicity vs model complexity

# Too simple → underfitting
# Too complex → overfitting

print("Choosing the right model complexity is important.")


# ==========================================================
# CELL 6 — UNDERFITTING
# ==========================================================

# Underfitting occurs when the model is too simple
# and cannot capture patterns in the data.

# Characteristics:
# High training error
# High testing error

print("Underfitting occurs when the model is too simple.")


# ==========================================================
# CELL 7 — OVERFITTING
# ==========================================================

# Overfitting occurs when the model memorizes training data.

# Characteristics:
# Very low training error
# High testing error

print("Overfitting occurs when the model learns noise.")


# ==========================================================
# CELL 8 — BIAS VS VARIANCE
# ==========================================================

# Bias:
# Error due to overly simple model assumptions.

# Variance:
# Error due to model sensitivity to training data.

# High Bias → Underfitting
# High Variance → Overfitting

print("Bias-Variance tradeoff is central to machine learning.")


# ==========================================================
# CELL 9 — BIAS VARIANCE DEMO
# ==========================================================

# Simple example showing prediction variance

data = np.array([2,4,6,8,10])

mean_prediction = np.mean(data)

variance = np.var(data)

print("Mean prediction:", mean_prediction)
print("Variance:", variance)


# ==========================================================
# CELL 10 — WHY THIS MATTERS IN DEEP LEARNING
# ==========================================================

# Deep neural networks have millions of parameters.

# Without proper techniques:
# - Overfitting becomes very common.

# To solve this we use:
# Regularization
# Dropout
# Data augmentation
# Early stopping

print("Controlling overfitting is critical in deep learning.")
```
