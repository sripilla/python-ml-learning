```python
# ==========================================================
# CELL 1 — WHAT IS REGULARIZATION
# ==========================================================

# Regularization techniques help prevent overfitting in deep learning models.

# When a neural network has too many parameters,
# it can memorize the training data instead of learning patterns.

# Regularization introduces constraints that help the model generalize better.

print("Regularization helps prevent overfitting in neural networks.")


# ==========================================================
# CELL 2 — L1 REGULARIZATION (LASSO)
# ==========================================================

# L1 regularization adds the absolute value of weights to the loss function.

# Formula:
# Loss = Original Loss + λ * Σ |w|

# This encourages sparsity (many weights become zero).

import numpy as np

weights = np.array([0.5, -0.3, 0.1, 0.8])

l1_penalty = np.sum(np.abs(weights))

print("L1 penalty:", l1_penalty)


# ==========================================================
# CELL 3 — L2 REGULARIZATION (RIDGE)
# ==========================================================

# L2 regularization adds squared weights to the loss.

# Formula:
# Loss = Original Loss + λ * Σ (w^2)

# This prevents weights from becoming very large.

l2_penalty = np.sum(weights ** 2)

print("L2 penalty:", l2_penalty)


# ==========================================================
# CELL 4 — EFFECT OF REGULARIZATION
# ==========================================================

# Regularization reduces model complexity.

# Benefits:
# - Reduces overfitting
# - Improves generalization
# - Controls weight magnitude

print("Regularization improves model generalization.")


# ==========================================================
# CELL 5 — EARLY STOPPING
# ==========================================================

# Early stopping is another technique to prevent overfitting.

# During training:
# - Training loss decreases continuously
# - Validation loss may start increasing

# When validation loss starts increasing,
# we stop training.

print("Early stopping stops training before overfitting occurs.")


# ==========================================================
# CELL 6 — WHAT IS DROPOUT
# ==========================================================

# Dropout is a regularization technique used in deep neural networks.

# During training:
# Some neurons are randomly turned OFF.

# This prevents neurons from relying too much on specific connections.

print("Dropout randomly disables neurons during training.")


# ==========================================================
# CELL 7 — DROPOUT DEMONSTRATION
# ==========================================================

# Example: randomly dropping neurons

neurons = np.array([1, 1, 1, 1, 1])

dropout_rate = 0.4

mask = np.random.binomial(1, 1 - dropout_rate, size=neurons.shape)

dropped_neurons = neurons * mask

print("Original neurons:", neurons)
print("Dropout mask:", mask)
print("After dropout:", dropped_neurons)


# ==========================================================
# CELL 8 — WHY DROPOUT WORKS
# ==========================================================

# Dropout forces the network to learn redundant representations.

# Advantages:
# - Prevents co-adaptation of neurons
# - Improves generalization
# - Reduces overfitting

print("Dropout makes neural networks more robust.")


# ==========================================================
# CELL 9 — OTHER REGULARIZATION METHODS
# ==========================================================

# Additional techniques used in deep learning:

# - Data augmentation
# - Batch normalization
# - Weight decay
# - Noise injection

print("Multiple regularization methods exist for deep learning.")


# ==========================================================
# CELL 10 — SUMMARY
# ==========================================================

# Key techniques to prevent overfitting:

# 1. L1 Regularization
# 2. L2 Regularization
# 3. Early Stopping
# 4. Dropout
# 5. Data Augmentation

print("Regularization is essential for training deep neural networks.")
```
