```python
# ==========================================================
# CELL 1 — INTRODUCTION TO OPTIMIZATION
# ==========================================================

# Optimization algorithms are used to update the weights
# of a neural network in order to minimize the loss function.

# During training:
# The model makes predictions
# The loss is computed
# Gradients are calculated
# The optimizer updates the weights

print("Optimization algorithms help neural networks learn efficiently.")


# ==========================================================
# CELL 2 — GRADIENT DESCENT
# ==========================================================

# Gradient Descent is the most basic optimization algorithm.

# Formula:
# weight_new = weight_old - learning_rate * gradient

learning_rate = 0.01
weight = 0.5
gradient = 0.2

updated_weight = weight - learning_rate * gradient

print("Updated weight:", updated_weight)


# ==========================================================
# CELL 3 — TYPES OF GRADIENT DESCENT
# ==========================================================

# 1. Batch Gradient Descent
# Uses the entire dataset for one update.

# 2. Stochastic Gradient Descent (SGD)
# Updates weights after each training example.

# 3. Mini-batch Gradient Descent
# Uses small batches of data.

print("Mini-batch gradient descent is most commonly used.")


# ==========================================================
# CELL 4 — STOCHASTIC GRADIENT DESCENT (SGD)
# ==========================================================

import numpy as np

weights = np.array([0.5, -0.3])
gradients = np.array([0.1, -0.2])
learning_rate = 0.01

weights = weights - learning_rate * gradients

print("Updated weights using SGD:", weights)


# ==========================================================
# CELL 5 — MOMENTUM OPTIMIZER
# ==========================================================

# Momentum helps accelerate gradient descent.

# It remembers past gradients to smooth updates.

velocity = np.array([0.0, 0.0])
beta = 0.9

velocity = beta * velocity + (1 - beta) * gradients
weights = weights - learning_rate * velocity

print("Weights after momentum update:", weights)


# ==========================================================
# CELL 6 — RMSPROP
# ==========================================================

# RMSprop adapts the learning rate
# by dividing gradients by their moving average.

epsilon = 1e-8
avg_sq_grad = gradients**2

weights = weights - learning_rate * gradients / (np.sqrt(avg_sq_grad) + epsilon)

print("Weights after RMSprop step:", weights)


# ==========================================================
# CELL 7 — ADAM OPTIMIZER
# ==========================================================

# Adam combines momentum and RMSprop.

# It is one of the most popular optimizers in deep learning.

m = np.zeros_like(weights)
v = np.zeros_like(weights)

beta1 = 0.9
beta2 = 0.999

m = beta1 * m + (1 - beta1) * gradients
v = beta2 * v + (1 - beta2) * (gradients ** 2)

weights = weights - learning_rate * m / (np.sqrt(v) + epsilon)

print("Weights after Adam update:", weights)


# ==========================================================
# CELL 8 — PARAMETER INITIALIZATION
# ==========================================================

# Proper weight initialization helps training stability.

# Common methods:
# - Random initialization
# - Xavier initialization
# - He initialization

weights = np.random.randn(3,3)

print("Randomly initialized weights:\n", weights)


# ==========================================================
# CELL 9 — CHALLENGES IN OPTIMIZATION
# ==========================================================

# Training deep neural networks can face problems such as:

# - Vanishing gradients
# - Exploding gradients
# - Slow convergence

print("Advanced optimizers help overcome these training challenges.")


# ==========================================================
# CELL 10 — SUMMARY
# ==========================================================

# Important optimizers:

# 1. Gradient Descent
# 2. SGD
# 3. Momentum
# 4. RMSprop
# 5. Adam

print("Adam optimizer is widely used in modern deep learning.")
```
