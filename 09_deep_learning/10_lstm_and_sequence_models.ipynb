```python
# ==========================================================
# CELL 1 — LIMITATIONS OF BASIC RNN
# ==========================================================

# Basic RNNs struggle with long sequences because of the
# vanishing gradient problem.

# When the sequence is long, earlier information is lost.

# Example:
# In long sentences, early words become hard to remember.

print("Basic RNNs struggle to remember long-term dependencies.")


# ==========================================================
# CELL 2 — INTRODUCTION TO LSTM
# ==========================================================

# LSTM stands for Long Short-Term Memory.

# It is a special type of RNN designed to remember
# information for long periods.

# LSTM solves the vanishing gradient problem using
# a memory cell and gating mechanisms.

print("LSTM networks can remember long-term information.")


# ==========================================================
# CELL 3 — LSTM ARCHITECTURE
# ==========================================================

# LSTM contains three main gates:

# 1. Forget Gate
#    Decides what information to discard.

# 2. Input Gate
#    Decides what new information to store.

# 3. Output Gate
#    Decides what information to output.

print("LSTM uses gates to control information flow.")


# ==========================================================
# CELL 4 — SIMPLE MEMORY UPDATE DEMO
# ==========================================================

import numpy as np

previous_memory = 0.6
new_information = 0.8

forget_gate = 0.7
input_gate = 0.5

memory = forget_gate * previous_memory + input_gate * new_information

print("Updated memory state:", memory)


# ==========================================================
# CELL 5 — BIDIRECTIONAL RNN
# ==========================================================

# Bidirectional RNN processes sequences in both directions.

# Forward direction:
# past → future

# Backward direction:
# future → past

# This helps the model understand context from both sides.

print("Bidirectional RNN improves context understanding.")


# ==========================================================
# CELL 6 — ENCODER–DECODER ARCHITECTURE
# ==========================================================

# Encoder–Decoder models are widely used for sequence tasks.

# Example:
# Language translation

# Encoder:
# Converts input sequence into a context vector.

# Decoder:
# Generates the output sequence.

print("Encoder–Decoder models are used in translation systems.")


# ==========================================================
# CELL 7 — SEQUENCE-TO-SEQUENCE MODEL
# ==========================================================

# Sequence-to-sequence (Seq2Seq) models map
# an input sequence to an output sequence.

# Example:
# English → French translation

print("Seq2Seq models transform one sequence into another.")


# ==========================================================
# CELL 8 — SIMPLE SEQUENCE PROCESSING EXAMPLE
# ==========================================================

sequence = [1, 2, 3, 4]

hidden_state = 0

for value in sequence:
    hidden_state = np.tanh(value + hidden_state)

print("Final hidden state:", hidden_state)


# ==========================================================
# CELL 9 — APPLICATIONS OF LSTM
# ==========================================================

# LSTM networks are used in:

# - Speech recognition
# - Language translation
# - Text generation
# - Video analysis
# - Financial time-series prediction

print("LSTM networks are powerful for sequence learning tasks.")


# ==========================================================
# CELL 10 — SUMMARY
# ==========================================================

# Key sequence models:

# - RNN
# - Bidirectional RNN
# - LSTM
# - Encoder–Decoder

print("LSTM is one of the most important deep learning models.")
```
