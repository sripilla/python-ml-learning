```python
# ==========================================================
# CELL 1 — INTRODUCTION TO AUTOENCODERS
# ==========================================================

# Autoencoders are neural networks used for learning
# compressed representations of data.

# They are mainly used for:
# - Dimensionality reduction
# - Feature learning
# - Image denoising
# - Data compression

print("Autoencoders learn compact representations of data.")


# ==========================================================
# CELL 2 — AUTOENCODER ARCHITECTURE
# ==========================================================

# Autoencoders have two main parts:

# 1. Encoder
#    Compresses input into a lower-dimensional representation.

# 2. Decoder
#    Reconstructs the original input from the compressed representation.

print("Autoencoders contain encoder and decoder networks.")


# ==========================================================
# CELL 3 — SIMPLE AUTOENCODER CONCEPT
# ==========================================================

import numpy as np

# example input data
x = np.array([1, 2, 3])

# encoder compresses the data
encoded = np.mean(x)

# decoder reconstructs data
decoded = np.array([encoded, encoded, encoded])

print("Encoded representation:", encoded)
print("Decoded output:", decoded)


# ==========================================================
# CELL 4 — TYPES OF AUTOENCODERS
# ==========================================================

# Common types of autoencoders:

# - Sparse Autoencoder
# - Denoising Autoencoder
# - Variational Autoencoder (VAE)

print("Different types of autoencoders exist for different tasks.")


# ==========================================================
# CELL 5 — AUTOENCODER APPLICATIONS
# ==========================================================

# Applications include:

# - Image denoising
# - Feature extraction
# - Anomaly detection
# - Data compression

print("Autoencoders are useful for unsupervised learning tasks.")


# ==========================================================
# CELL 6 — INTRODUCTION TO GANs
# ==========================================================

# GAN stands for Generative Adversarial Network.

# GANs are used to generate new data
# that resembles real data.

print("GANs are powerful generative models.")


# ==========================================================
# CELL 7 — GAN ARCHITECTURE
# ==========================================================

# GAN consists of two networks:

# 1. Generator
#    Generates fake samples.

# 2. Discriminator
#    Tries to distinguish between real and fake samples.

print("GAN training is a competition between generator and discriminator.")


# ==========================================================
# CELL 8 — SIMPLE GAN CONCEPT DEMO
# ==========================================================

# Example simulation

real_data = np.array([1,2,3,4])

# generator produces fake data
fake_data = np.random.choice(real_data)

print("Generated fake sample:", fake_data)


# ==========================================================
# CELL 9 — GAN APPLICATIONS
# ==========================================================

# GAN applications include:

# - Image generation
# - Deepfakes
# - Image-to-image translation
# - Super resolution
# - Art generation

print("GANs can generate realistic synthetic data.")


# ==========================================================
# CELL 10 — SUMMARY
# ==========================================================

# Important generative models:

# - Autoencoders
# - Variational Autoencoders
# - Generative Adversarial Networks

print("Generative models create new data similar to training data.")
```
