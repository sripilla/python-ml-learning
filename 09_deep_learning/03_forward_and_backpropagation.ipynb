```python
# ==========================================================
# CELL 1 — FORWARD PROPAGATION
# ==========================================================

# Forward propagation is the process where input data
# passes through the neural network to produce an output.

# Steps:
# 1. Input data enters the network
# 2. Each layer computes weighted sums
# 3. Activation function is applied
# 4. Final output is produced

print("Forward propagation computes predictions from inputs.")


# ==========================================================
# CELL 2 — SIMPLE FORWARD PROPAGATION EXAMPLE
# ==========================================================

import numpy as np

# input vector
X = np.array([1, 2])

# weights
W = np.array([[0.5, 0.2],
              [0.3, 0.8]])

# bias
b = np.array([0.1, 0.2])

# linear transformation
Z = np.dot(X, W) + b

print("Linear output:", Z)


# ==========================================================
# CELL 3 — ACTIVATION FUNCTION
# ==========================================================

# Activation functions introduce non-linearity
# allowing neural networks to learn complex patterns.

def relu(x):
    return np.maximum(0, x)

A = relu(Z)

print("Activation output:", A)


# ==========================================================
# CELL 4 — COMPUTATIONAL GRAPHS
# ==========================================================

# A computational graph represents mathematical operations
# as nodes and edges.

# Example:
# y = (a + b) * c

a = 2
b = 3
c = 4

y = (a + b) * c

print("Output:", y)

# In deep learning frameworks (TensorFlow, PyTorch)
# these graphs are used to automatically compute gradients.


# ==========================================================
# CELL 5 — LOSS FUNCTION
# ==========================================================

# After forward propagation, we compute the loss
# to measure prediction error.

# Example: Mean Squared Error (MSE)

y_true = 10
y_pred = y

loss = (y_true - y_pred) ** 2

print("Loss:", loss)


# ==========================================================
# CELL 6 — WHAT IS BACKPROPAGATION
# ==========================================================

# Backpropagation is the algorithm used to update
# neural network weights by computing gradients.

# Steps:
# 1. Compute loss
# 2. Calculate gradients using chain rule
# 3. Update weights using gradient descent

print("Backpropagation adjusts weights to reduce prediction error.")


# ==========================================================
# CELL 7 — GRADIENT DESCENT
# ==========================================================

# Gradient descent minimizes the loss function.

# Formula:
# new_weight = old_weight - learning_rate * gradient

learning_rate = 0.01
weight = 0.5
gradient = 0.2

new_weight = weight - learning_rate * gradient

print("Updated weight:", new_weight)


# ==========================================================
# CELL 8 — SIMPLE BACKPROPAGATION DEMO
# ==========================================================

# simple linear model
w = 0.5
x = 2
y_true = 4

# forward pass
y_pred = w * x

# loss
loss = (y_true - y_pred) ** 2

# gradient
gradient = -2 * x * (y_true - y_pred)

# update weight
learning_rate = 0.01
w = w - learning_rate * gradient

print("Updated weight after backprop:", w)


# ==========================================================
# CELL 9 — WHY BACKPROPAGATION IS IMPORTANT
# ==========================================================

# Backpropagation enables neural networks to learn.

# Without backpropagation:
# Deep learning would not work.

# It efficiently computes gradients for millions
# of parameters in deep neural networks.

print("Backpropagation is the core training algorithm of deep learning.")
```
