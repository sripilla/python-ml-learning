```python
# ==========================================================
# CELL 1 — INTRODUCTION TO RECURRENT NEURAL NETWORKS (RNN)
# ==========================================================

# Recurrent Neural Networks are designed for sequential data.

# Unlike feedforward neural networks,
# RNNs have connections that form cycles.

# This allows them to maintain a memory of previous inputs.

# Examples of sequential data:
# - Text
# - Speech
# - Time series
# - Video frames

print("RNNs are designed to process sequential data.")


# ==========================================================
# CELL 2 — WHY RNNs ARE NEEDED
# ==========================================================

# Traditional neural networks treat inputs independently.

# However many problems require context.

# Example sentence:
# "The movie was not good"

# The meaning depends on previous words.

print("RNNs capture relationships between sequential inputs.")


# ==========================================================
# CELL 3 — RNN ARCHITECTURE
# ==========================================================

# Basic RNN structure:

# Input → Hidden State → Output

# Hidden state stores information from previous time steps.

print("Hidden states allow RNNs to remember past information.")


# ==========================================================
# CELL 4 — SIMPLE RNN COMPUTATION
# ==========================================================

import numpy as np

# input at time t
x_t = 2

# previous hidden state
h_prev = 0.5

# weights
W_x = 0.8
W_h = 0.3

# new hidden state
h_t = np.tanh(W_x * x_t + W_h * h_prev)

print("New hidden state:", h_t)


# ==========================================================
# CELL 5 — UNFOLDING RNN THROUGH TIME
# ==========================================================

# RNNs process sequences step-by-step.

# Example sequence:
# x1 → x2 → x3 → x4

# Each step updates the hidden state.

print("RNNs are often visualized as networks unfolded through time.")


# ==========================================================
# CELL 6 — SEQUENCE MODELING
# ==========================================================

# RNNs can model different sequence tasks:

# 1. One-to-one
#    Image classification

# 2. One-to-many
#    Image captioning

# 3. Many-to-one
#    Sentiment analysis

# 4. Many-to-many
#    Language translation

print("RNNs support multiple sequence architectures.")


# ==========================================================
# CELL 7 — TRAINING RNNs
# ==========================================================

# RNNs are trained using Backpropagation Through Time (BPTT).

# The network is unfolded across time
# and gradients are computed across all time steps.

print("RNNs are trained using Backpropagation Through Time.")


# ==========================================================
# CELL 8 — VANISHING GRADIENT PROBLEM
# ==========================================================

# During training, gradients can become extremely small.

# This prevents early layers from learning.

# This problem is called the Vanishing Gradient Problem.

print("Vanishing gradients make training long sequences difficult.")


# ==========================================================
# CELL 9 — EXPLODING GRADIENT PROBLEM
# ==========================================================

# Sometimes gradients become extremely large.

# This causes unstable training.

# Techniques like gradient clipping help control this problem.

print("Exploding gradients can destabilize neural network training.")


# ==========================================================
# CELL 10 — APPLICATIONS OF RNNs
# ==========================================================

# RNN applications include:

# - Speech recognition
# - Language translation
# - Text generation
# - Music generation
# - Time-series forecasting

print("RNNs are widely used for sequential learning tasks.")
```
