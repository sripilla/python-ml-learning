```python
# ==========================================================
# CELL 1 — INTRODUCTION TO DEEP LEARNING
# ==========================================================

# Deep Learning is a subset of Machine Learning that uses
# Artificial Neural Networks with many layers.

# These networks can automatically learn patterns from large datasets.

# Examples of Deep Learning applications:
# - Image Recognition
# - Speech Recognition
# - Natural Language Processing
# - Autonomous Driving
# - Medical Diagnosis

print("Deep Learning allows computers to learn complex patterns from data.")


# ==========================================================
# CELL 2 — WHY DEEP LEARNING?
# ==========================================================

# Traditional Machine Learning:
# Humans manually design features.

# Deep Learning:
# The model automatically learns features from raw data.

# Example:
# Image classification

# ML approach:
# Handcraft edges, textures, shapes.

# DL approach:
# Neural networks automatically learn these features.


# ==========================================================
# CELL 3 — HISTORY OF DEEP LEARNING
# ==========================================================

# 1957 — Perceptron (Frank Rosenblatt)
# 1986 — Backpropagation popularized
# 1998 — LeNet CNN by Yann LeCun
# 2012 — AlexNet revolutionized deep learning

print("Modern Deep Learning started booming after 2012.")


# ==========================================================
# CELL 4 — WHAT IS A TENSOR?
# ==========================================================

# A Tensor is a generalization of vectors and matrices.

# Tensor dimensions:

# Scalar → 0D tensor
# Vector → 1D tensor
# Matrix → 2D tensor
# Higher dimensions → ND tensor

import numpy as np

scalar = np.array(5)
vector = np.array([1,2,3])
matrix = np.array([[1,2],[3,4]])

print("Scalar:", scalar)
print("Vector:", vector)
print("Matrix:\n", matrix)


# ==========================================================
# CELL 5 — TENSOR SHAPES
# ==========================================================

print("Scalar shape:", scalar.shape)
print("Vector shape:", vector.shape)
print("Matrix shape:", matrix.shape)


# ==========================================================
# CELL 6 — HIGHER DIMENSION TENSORS
# ==========================================================

tensor3d = np.random.rand(3,4,5)

print("3D Tensor shape:", tensor3d.shape)


# ==========================================================
# CELL 7 — BASIC TENSOR OPERATIONS
# ==========================================================

A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

# addition
print("Addition:\n", A + B)

# multiplication
print("Elementwise multiplication:\n", A * B)

# matrix multiplication
print("Matrix multiplication:\n", A @ B)


# ==========================================================
# CELL 8 — WHY TENSORS ARE IMPORTANT
# ==========================================================

# Deep Learning frameworks like:
# TensorFlow
# PyTorch

# represent all data as tensors.

# Examples:
# Image → 3D tensor
# Batch of images → 4D tensor
# Video → 5D tensor

print("Tensors are the basic data structure used in deep learning.")
```
