```python
# ============================================
# CELL 1 - SUPPORT VECTOR MACHINE INTRODUCTION
# ============================================

"""
Support Vector Machine (SVM) is a supervised Machine Learning algorithm
used for classification and regression.

Main Idea:
SVM tries to find the BEST boundary (called a hyperplane)
that separates different classes.

Example:

PASS   PASS   PASS
--------------------  ← Hyperplane
FAIL   FAIL   FAIL

The best hyperplane is the one that has the
MAXIMUM MARGIN from both classes.

Key Concepts:

Hyperplane → Decision boundary
Margin → Distance between hyperplane and nearest data points
Support Vectors → Data points closest to the boundary
"""

print("Support Vector Machine (SVM)")


# ============================================
# CELL 2 - IMPORT LIBRARIES
# ============================================

"""
Libraries used:

NumPy → numerical operations
Pandas → dataset handling
Matplotlib → visualization
Scikit-Learn → SVM model
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


# ============================================
# CELL 3 - CREATE SAMPLE DATASET
# ============================================

"""
Example dataset:

Hours studied vs pass/fail
"""

data = {
    "Hours_Studied": [1,2,3,4,5,6,7,8],
    "Pass": [0,0,0,0,1,1,1,1]
}

df = pd.DataFrame(data)

print(df)


# ============================================
# CELL 4 - FEATURES AND TARGET
# ============================================

"""
Feature → Hours Studied
Target → Pass / Fail
"""

X = df[["Hours_Studied"]]
y = df["Pass"]

print("Features:")
print(X)

print("\nTarget:")
print(y)


# ============================================
# CELL 5 - TRAIN TEST SPLIT
# ============================================

"""
Split dataset into training and testing data
"""

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

print("Training samples:", len(X_train))
print("Testing samples:", len(X_test))


# ============================================
# CELL 6 - TRAIN SVM MODEL
# ============================================

"""
Create SVM classifier.

kernel = linear (simple boundary)
Other kernels:
rbf
poly
sigmoid
"""

model = SVC(kernel='linear')

model.fit(X_train, y_train)

print("SVM model trained")


# ============================================
# CELL 7 - MAKE PREDICTIONS
# ============================================

"""
Predict classes for test dataset
"""

predictions = model.predict(X_test)

print("Actual values:", list(y_test))
print("Predicted values:", predictions)


# ============================================
# CELL 8 - MODEL ACCURACY
# ============================================

"""
Evaluate model performance
"""

accuracy = accuracy_score(y_test, predictions)

print("Model Accuracy:", accuracy)


# ============================================
# CELL 9 - PREDICT NEW VALUE
# ============================================

"""
Predict if a student studying 5 hours
will pass or fail
"""

new_student = np.array([[5]])

prediction = model.predict(new_student)

if prediction[0] == 1:
    print("Prediction: PASS")
else:
    print("Prediction: FAIL")


# ============================================
# CELL 10 - VISUALIZE DATA
# ============================================

"""
Visualize dataset points
"""

plt.scatter(df["Hours_Studied"], df["Pass"])

plt.xlabel("Hours Studied")
plt.ylabel("Pass / Fail")
plt.title("SVM Dataset")

plt.show()


# ============================================
# CELL 11 - SUMMARY
# ============================================

"""
Summary:

✔ SVM finds the best decision boundary
✔ Uses hyperplanes to separate classes
✔ Maximizes margin between classes
✔ Works well with high-dimensional data
✔ Can handle linear and non-linear classification
"""

print("SVM notebook completed")


"""
What You Learned

Key concepts:

    Support Vector Machines

    Hyperplane

    Margin

    Support vectors

    Kernel trick

    Classification using SVM
"""
```
