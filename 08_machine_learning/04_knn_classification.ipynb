```python
# ============================================
# CELL 1 - K NEAREST NEIGHBORS INTRODUCTION
# ============================================

"""
K-Nearest Neighbors (KNN) is a simple Machine Learning algorithm
used for both classification and regression.

It works based on similarity between data points.

Idea:
When a new data point comes, the algorithm finds the
"K nearest neighbors" in the dataset and assigns the
most common class among them.

Example:
If 5 nearest neighbors contain
3 PASS and 2 FAIL → prediction = PASS

Distance is usually calculated using:

Euclidean Distance

distance = sqrt((x1-x2)^2 + (y1-y2)^2)
"""

print("K-Nearest Neighbors Algorithm")


# ============================================
# CELL 2 - IMPORT LIBRARIES
# ============================================

"""
Libraries used:

NumPy → numerical operations
Pandas → dataset handling
Matplotlib → visualization
Scikit-Learn → KNN model
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


# ============================================
# CELL 3 - CREATE SAMPLE DATASET
# ============================================

"""
Example Problem:

Predict whether a student PASSES or FAILS
based on hours studied.

0 → Fail
1 → Pass
"""

data = {
    "Hours_Studied": [1,2,3,4,5,6,7,8],
    "Pass": [0,0,0,0,1,1,1,1]
}

df = pd.DataFrame(data)

print(df)


# ============================================
# CELL 4 - DEFINE FEATURES AND TARGET
# ============================================

"""
Feature → Hours Studied
Target → Pass / Fail
"""

X = df[["Hours_Studied"]]
y = df["Pass"]

print("Features:")
print(X)

print("\nTarget:")
print(y)


# ============================================
# CELL 5 - SPLIT DATASET
# ============================================

"""
Split dataset into:

Training data
Testing data
"""

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

print("Training samples:", len(X_train))
print("Testing samples:", len(X_test))


# ============================================
# CELL 6 - CREATE KNN MODEL
# ============================================

"""
n_neighbors = number of neighbors (K)

Common values: 3, 5, 7
"""

knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X_train, y_train)

print("KNN model trained")


# ============================================
# CELL 7 - MAKE PREDICTIONS
# ============================================

"""
Predict class of test data
"""

predictions = knn.predict(X_test)

print("Actual values:", list(y_test))
print("Predicted values:", predictions)


# ============================================
# CELL 8 - MODEL ACCURACY
# ============================================

"""
Calculate prediction accuracy
"""

accuracy = accuracy_score(y_test, predictions)

print("Model Accuracy:", accuracy)


# ============================================
# CELL 9 - PREDICT NEW DATA
# ============================================

"""
Predict if a student studying 5 hours
will pass or fail
"""

new_student = np.array([[5]])

prediction = knn.predict(new_student)

if prediction[0] == 1:
    print("Prediction: PASS")
else:
    print("Prediction: FAIL")


# ============================================
# CELL 10 - VISUALIZE DATA
# ============================================

"""
Visualize dataset
"""

plt.scatter(df["Hours_Studied"], df["Pass"])

plt.xlabel("Hours Studied")
plt.ylabel("Pass / Fail")
plt.title("KNN Dataset")

plt.show()


# ============================================
# CELL 11 - SUMMARY
# ============================================

"""
Summary:

✔ KNN is a distance-based algorithm
✔ It looks at nearest neighbors
✔ Majority voting decides the class
✔ Choosing correct K is important
✔ Works well for small datasets
"""

print("KNN notebook completed")


"""
What You Learned

This notebook covers:

    Distance-based learning

    K nearest neighbors

    Majority voting

    Model training

    Accuracy evaluation
"""
```
