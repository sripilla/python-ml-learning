```python
# ============================================
# CELL 1 - FEATURE SCALING INTRODUCTION
# ============================================

"""
Feature Scaling is the process of bringing all features
to a similar range.

Why is it important?

Some Machine Learning algorithms depend on distance.

Example algorithms:
KNN
SVM
K-Means
Gradient Descent

Example problem:

Age = 25
Salary = 50000

Salary dominates the distance calculation.

Feature scaling fixes this problem.
"""

print("Feature Scaling in Machine Learning")


# ============================================
# CELL 2 - TYPES OF FEATURE SCALING
# ============================================

"""
Two common scaling techniques:

1️⃣ Standardization (Z-score scaling)

Formula:

z = (x - mean) / standard deviation

Output range:
Centered around 0

Used for:
SVM
Logistic Regression
Neural Networks


2️⃣ Normalization (Min-Max Scaling)

Formula:

x_scaled = (x - min) / (max - min)

Output range:
0 to 1
"""

print("Standardization and Normalization")


# ============================================
# CELL 3 - IMPORT LIBRARIES
# ============================================

import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler


# ============================================
# CELL 4 - CREATE SAMPLE DATASET
# ============================================

"""
Example dataset with features on different scales
"""

data = {
    "Age": [22, 25, 30, 35, 40],
    "Salary": [20000, 35000, 50000, 65000, 80000]
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)


# ============================================
# CELL 5 - STANDARDIZATION
# ============================================

"""
Apply StandardScaler
"""

scaler = StandardScaler()

standardized_data = scaler.fit_transform(df)

df_standardized = pd.DataFrame(
    standardized_data,
    columns=df.columns
)

print("\nStandardized Data:")
print(df_standardized)


# ============================================
# CELL 6 - NORMALIZATION
# ============================================

"""
Apply MinMaxScaler
"""

normalizer = MinMaxScaler()

normalized_data = normalizer.fit_transform(df)

df_normalized = pd.DataFrame(
    normalized_data,
    columns=df.columns
)

print("\nNormalized Data:")
print(df_normalized)


# ============================================
# CELL 7 - COMPARISON
# ============================================

"""
Compare original vs scaled data
"""

print("\nOriginal Data:")
print(df)

print("\nStandardized Data:")
print(df_standardized)

print("\nNormalized Data:")
print(df_normalized)


# ============================================
# CELL 8 - SUMMARY
# ============================================

"""
Summary:

✔ Feature scaling ensures features have similar ranges
✔ Prevents large-value features from dominating
✔ Standardization → mean = 0, std = 1
✔ Normalization → range between 0 and 1
✔ Important for distance-based algorithms
"""

print("Feature Scaling notebook completed")

"""
What You Learned

This notebook covers:

    Why Feature Scaling is necessary

    Standardization (Z-score scaling)

    Normalization (Min-Max scaling)

    How to use StandardScaler

    How to use MinMaxScaler
"""
```
