```python
# ============================================
# CELL 1 - K-MEANS CLUSTERING INTRODUCTION
# ============================================

"""
K-Means is an Unsupervised Machine Learning algorithm.

Unlike supervised learning, there are NO labels.

Goal:
Group similar data points into clusters.

Example:
Customer segmentation

Customers with similar buying behavior
will be grouped together.

K-Means Steps:

1. Choose number of clusters (K)
2. Randomly initialize cluster centers
3. Assign each point to nearest cluster
4. Update cluster centers
5. Repeat until convergence
"""

print("K-Means Clustering Algorithm")


# ============================================
# CELL 2 - IMPORT LIBRARIES
# ============================================

"""
Libraries used:

NumPy → numerical computation
Pandas → dataset handling
Matplotlib → visualization
Scikit-Learn → K-Means model
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans


# ============================================
# CELL 3 - CREATE SAMPLE DATASET
# ============================================

"""
Example dataset:
Customer age vs spending score
"""

data = {
    "Age": [22,25,47,52,46,56,23,27,48,54],
    "Spending": [25,30,70,65,68,72,28,35,75,80]
}

df = pd.DataFrame(data)

print(df)


# ============================================
# CELL 4 - VISUALIZE DATA
# ============================================

"""
Before clustering, visualize the dataset
"""

plt.scatter(df["Age"], df["Spending"])

plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.title("Customer Data")

plt.show()


# ============================================
# CELL 5 - APPLY K-MEANS
# ============================================

"""
Create K-Means model

n_clusters = number of clusters
"""

kmeans = KMeans(n_clusters=2, random_state=42)

clusters = kmeans.fit_predict(df)

print("Cluster assignments:")
print(clusters)


# ============================================
# CELL 6 - ADD CLUSTERS TO DATAFRAME
# ============================================

"""
Attach cluster labels to dataset
"""

df["Cluster"] = clusters

print(df)


# ============================================
# CELL 7 - VISUALIZE CLUSTERS
# ============================================

"""
Visualize clusters with colors
"""

plt.scatter(df["Age"], df["Spending"], c=df["Cluster"])

plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.title("K-Means Clustering")

plt.show()


# ============================================
# CELL 8 - CLUSTER CENTERS
# ============================================

"""
Cluster centers represent the average
point of each cluster
"""

centers = kmeans.cluster_centers_

print("Cluster Centers:")
print(centers)


# ============================================
# CELL 9 - VISUALIZE CENTERS
# ============================================

"""
Plot clusters and their centers
"""

plt.scatter(df["Age"], df["Spending"], c=df["Cluster"])
plt.scatter(centers[:,0], centers[:,1], s=200, marker='X')

plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.title("Clusters with Centers")

plt.show()


# ============================================
# CELL 10 - SUMMARY
# ============================================

"""
Summary:

✔ K-Means is an unsupervised algorithm
✔ It groups similar data points into clusters
✔ Each cluster has a center (centroid)
✔ Used in customer segmentation, image compression
"""

print("K-Means notebook completed")

"""
What You Learned

This notebook covers:

    Unsupervised learning

    Clustering

    K-Means algorithm

    Cluster centers

    Data segmentation
"""

```
