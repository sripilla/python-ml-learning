```python
"""
1. Element-wise Operations
2. Scalar Operations
3. Mathematical Functions
4. Aggregation Functions
5. Axis Operations
6. Broadcasting Basics
7. Practice Exercises
"""

# Import NumPy
import numpy as np


# Element-wise Operations
# NumPy performs operations element by element.

a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)
print(a - b)
print(a * b)
print(a / b)


# Scalar Operations
# Operations with a single number apply to all elements.

a = np.array([1,2,3,4])

print(a + 10)
print(a * 2)


# Mathematical Functions
# NumPy provides many mathematical functions.

arr = np.array([1,4,9,16])

print(np.sqrt(arr))
print(np.exp(arr))
print(np.log(arr))


# Aggregation Functions
# These functions summarize data.

arr = np.array([10,20,30,40,50])

print(np.sum(arr))
print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))
print(np.std(arr))


# Axis Operations
# For 2D arrays we can perform operations row-wise or column-wise.

matrix = np.array([[1,2,3],
                   [4,5,6]])

print(np.sum(matrix, axis=0))  # column sums
print(np.sum(matrix, axis=1))  # row sums


# Broadcasting Basics
# Broadcasting allows operations between arrays of different shapes.

a = np.array([1,2,3])
b = 2

print(a * b)


# Another broadcasting example

matrix = np.array([[1,2,3],
                   [4,5,6]])

vector = np.array([1,1,1])

print(matrix + vector)


# Practice Exercise 1
# Multiply two arrays element-wise.

a = np.array([2,4,6])
b = np.array([1,3,5])

print(a * b)


# Practice Exercise 2
# Find the mean of an array.

arr = np.array([5,10,15,20])

print(np.mean(arr))


# Practice Exercise 3
# Find column sums of a matrix.

matrix = np.array([[2,4,6],
                   [1,3,5]])

print(np.sum(matrix, axis=0))
```
