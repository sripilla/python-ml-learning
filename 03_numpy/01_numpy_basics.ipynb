```python
"""
1. What is NumPy
2. Installing and Importing NumPy
3. Creating NumPy Arrays
4. Array Dimensions
5. Array Data Types
6. Converting Lists to Arrays
7. Basic Array Operations
8. Practice Exercises
"""

# What is NumPy
# NumPy stands for Numerical Python.
# It is a library used for numerical computing in Python.
# NumPy provides fast multidimensional arrays and mathematical operations.

# Installing NumPy (if not installed)
# pip install numpy


# Importing NumPy

import numpy as np


# Creating NumPy Arrays
# Arrays can be created from Python lists.

arr = np.array([1, 2, 3, 4, 5])

print(arr)
print(type(arr))


# Creating a 2D Array

matrix = np.array([[1,2,3],
                   [4,5,6]])

print(matrix)


# Array Dimensions
# ndim returns the number of dimensions of an array.

print(arr.ndim)
print(matrix.ndim)


# Array Shape
# shape returns rows and columns.

print(matrix.shape)


# Array Data Types
# dtype shows the data type of elements.

print(arr.dtype)


# Converting Lists to Arrays

numbers_list = [10, 20, 30, 40]

numbers_array = np.array(numbers_list)

print(numbers_array)


# Basic Array Operations

a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)
print(a * b)
print(a - b)


# Scalar Operations

a = np.array([1,2,3])

print(a * 2)
print(a + 10)


# Practice Exercise 1
# Create a NumPy array of numbers from 1 to 5.

arr = np.array([1,2,3,4,5])

print(arr)


# Practice Exercise 2
# Create a 2x2 matrix.

matrix = np.array([[1,2],
                   [3,4]])

print(matrix)


# Practice Exercise 3
# Perform element-wise addition.

a = np.array([1,2,3])
b = np.array([10,20,30])

print(a + b)
```
