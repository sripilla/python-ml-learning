```python
"""
1. Creating Arrays with zeros()
2. Creating Arrays with ones()
3. Creating Arrays with arange()
4. Creating Arrays with linspace()
5. Identity Matrix
6. Random Arrays
7. Practice Exercises
"""

# Import NumPy

import numpy as np


# Creating Arrays with zeros()
# zeros() creates an array filled with 0s.

arr = np.zeros(5)

print(arr)


# Creating a 2D array of zeros

matrix = np.zeros((3,3))

print(matrix)


# Creating Arrays with ones()
# ones() creates an array filled with 1s.

arr = np.ones(5)

print(arr)


# Creating a 2D array of ones

matrix = np.ones((2,4))

print(matrix)


# Creating Arrays with arange()
# arange() creates numbers within a range.

arr = np.arange(0,10)

print(arr)


# arange with step

arr = np.arange(0,20,2)

print(arr)


# Creating Arrays with linspace()
# linspace() creates evenly spaced numbers between two values.

arr = np.linspace(0,10,5)

print(arr)


# Identity Matrix
# eye() creates an identity matrix.

matrix = np.eye(3)

print(matrix)


# Random Arrays
# NumPy can generate random numbers.

random_array = np.random.rand(5)

print(random_array)


# Random 2D array

random_matrix = np.random.rand(3,3)

print(random_matrix)


# Random integers

rand_int = np.random.randint(1,10,5)

print(rand_int)


# Practice Exercise 1
# Create an array of 10 zeros.

arr = np.zeros(10)

print(arr)


# Practice Exercise 2
# Create numbers from 5 to 50 with step 5.

arr = np.arange(5,51,5)

print(arr)


# Practice Exercise 3
# Create an identity matrix of size 4.

matrix = np.eye(4)

print(matrix)
```
