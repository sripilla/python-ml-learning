```python
"""
1. Statistical Functions
2. Mean, Median, Standard Deviation
3. Finding Max and Min
4. Matrix Multiplication
5. Dot Product
6. Transpose of Matrix
7. Determinant
8. Practice Exercises
"""

# Import NumPy

import numpy as np


# Statistical Functions
# NumPy provides many functions to analyze data.

arr = np.array([10,20,30,40,50])

print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))


# Finding Maximum and Minimum

arr = np.array([5,15,25,35,45])

print(np.max(arr))
print(np.min(arr))


# Matrix Multiplication
# matmul() performs matrix multiplication.

A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])

result = np.matmul(A,B)

print(result)


# Dot Product
# dot() calculates dot product of two vectors.

a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.dot(a,b))


# Transpose of Matrix
# transpose() flips rows and columns.

matrix = np.array([[1,2,3],
                   [4,5,6]])

print(matrix.T)


# Determinant
# NumPy can compute determinant of a matrix.

matrix = np.array([[1,2],
                   [3,4]])

print(np.linalg.det(matrix))


# Practice Exercise 1
# Find the mean of an array.

arr = np.array([2,4,6,8,10])

print(np.mean(arr))


# Practice Exercise 2
# Compute dot product of two arrays.

a = np.array([1,3,5])
b = np.array([2,4,6])

print(np.dot(a,b))


# Practice Exercise 3
# Find transpose of a matrix.

matrix = np.array([[1,2],
                   [3,4]])

print(matrix.T)
```
