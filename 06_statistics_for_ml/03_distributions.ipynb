```python
# =========================================
# CELL 1 - TITLE AND INTRODUCTION
# =========================================
"""
DATA DISTRIBUTIONS

A distribution shows how values in a dataset are spread.

Important distributions in statistics and ML:

• Normal Distribution (Gaussian)
• Uniform Distribution
• Skewed Distribution

Understanding distributions helps:
- detect patterns
- identify outliers
- understand dataset behavior
"""


# =========================================
# CELL 2 - IMPORT LIBRARIES
# =========================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# =========================================
# CELL 3 - GENERATE NORMAL DISTRIBUTION DATA
# =========================================
data = np.random.normal(loc=50, scale=10, size=1000)

data[:10]


# =========================================
# CELL 4 - HISTOGRAM OF NORMAL DISTRIBUTION
# =========================================
plt.hist(data, bins=30)

plt.title("Histogram of Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()


# =========================================
# CELL 5 - KDE PLOT
# =========================================
sns.kdeplot(data, fill=True)

plt.title("Kernel Density Estimation")

plt.show()


# =========================================
# CELL 6 - UNIFORM DISTRIBUTION
# =========================================
uniform_data = np.random.uniform(0, 1, 1000)

plt.hist(uniform_data, bins=30)

plt.title("Uniform Distribution")

plt.show()


# =========================================
# CELL 7 - RIGHT SKEWED DISTRIBUTION
# =========================================
skewed_data = np.random.exponential(scale=2, size=1000)

plt.hist(skewed_data, bins=30)

plt.title("Right Skewed Distribution")

plt.show()


# =========================================
# CELL 8 - BOXPLOT OF DISTRIBUTION
# =========================================
sns.boxplot(x=data)

plt.title("Boxplot of Normal Distribution")

plt.show()


# =========================================
# CELL 9 - COMPARING DISTRIBUTIONS
# =========================================
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(data, bins=30)
plt.title("Normal")

plt.subplot(1,2,2)
plt.hist(skewed_data, bins=30)
plt.title("Skewed")

plt.show()


# =========================================
# CELL 10 - PRACTICE EXERCISES
# =========================================
"""
TRY THESE EXERCISES

1. Generate 500 numbers using normal distribution
2. Plot histogram of the dataset
3. Generate uniform distribution data
4. Compare two distributions using subplots
"""
```
