```python
# =========================================
# CELL 1 - TITLE AND INTRODUCTION
# =========================================
"""
CORRELATION AND COVARIANCE

These concepts help measure relationships between variables.

Covariance:
Measures how two variables change together.

Correlation:
Measures the strength and direction of the relationship
between two variables.

Correlation values range from:
-1  → strong negative relationship
 0  → no relationship
+1  → strong positive relationship
"""


# =========================================
# CELL 2 - IMPORT LIBRARIES
# =========================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# =========================================
# CELL 3 - CREATE SAMPLE DATA
# =========================================
data = {
    "Hours_Studied": [1,2,3,4,5,6,7,8],
    "Exam_Score": [35,40,50,55,65,70,80,90]
}

df = pd.DataFrame(data)

df


# =========================================
# CELL 4 - SCATTER PLOT (RELATIONSHIP)
# =========================================
plt.scatter(df["Hours_Studied"], df["Exam_Score"])

plt.title("Hours Studied vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")

plt.show()


# =========================================
# CELL 5 - COVARIANCE
# =========================================
cov_value = np.cov(df["Hours_Studied"], df["Exam_Score"])

print("Covariance Matrix:\n", cov_value)


# =========================================
# CELL 6 - CORRELATION
# =========================================
corr_value = df.corr()

print("Correlation Matrix:\n", corr_value)


# =========================================
# CELL 7 - CORRELATION VALUE
# =========================================
corr = df["Hours_Studied"].corr(df["Exam_Score"])

print("Correlation between Hours Studied and Exam Score:", corr)


# =========================================
# CELL 8 - HEATMAP
# =========================================
sns.heatmap(corr_value, annot=True)

plt.title("Correlation Heatmap")

plt.show()


# =========================================
# CELL 9 - RANDOM DATA CORRELATION EXAMPLE
# =========================================
x = np.random.rand(100)
y = x * 3 + np.random.randn(100)

plt.scatter(x,y)

plt.title("Positive Correlation Example")

plt.show()


# =========================================
# CELL 10 - PRACTICE EXERCISES
# =========================================
"""
TRY THESE EXERCISES

1. Create a dataset with three variables
2. Calculate correlation matrix
3. Plot heatmap of correlations
4. Visualize relationships using scatter plots
"""
```
