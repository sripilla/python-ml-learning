```python
# ==========================================================
# DATA PREPROCESSING – FEATURE ENGINEERING BASICS
# ==========================================================
# Feature Engineering is the process of creating new input
# variables (features) from existing data to improve the
# performance of machine learning models.
#
# In many real-world ML projects, feature engineering is one
# of the most important steps.
#
# A good feature can significantly improve model accuracy,
# sometimes even more than changing the ML algorithm itself.
#
# Common Feature Engineering Techniques:
#
# 1. Feature Creation
#    Creating new features from existing ones.
#    Example:
#    If we have "Height" and "Weight", we can create "BMI".
#
# 2. Feature Transformation
#    Changing the scale or distribution of features.
#    Example:
#    Log transformation of income data.
#
# 3. Feature Selection
#    Selecting only important features and removing
#    irrelevant ones.
#
# In this notebook we will demonstrate simple examples of
# feature engineering using pandas.
# ==========================================================



# ==========================================================
# CELL 1 – IMPORT LIBRARIES
# ==========================================================
# pandas → for dataset manipulation
# numpy → for numerical operations

import pandas as pd
import numpy as np



# ==========================================================
# CELL 2 – CREATE SAMPLE DATASET
# ==========================================================
# This dataset contains information about students.

data = {
    "Study_Hours": [2, 3, 5, 7, 8],
    "Sleep_Hours": [8, 7, 6, 5, 5],
    "Attendance": [70, 75, 85, 90, 95],
    "Exam_Score": [50, 55, 70, 80, 90]
}

df = pd.DataFrame(data)

df



# ==========================================================
# CELL 3 – FEATURE CREATION
# ==========================================================
# Here we create a new feature called "Productivity_Index".
#
# This combines study hours and attendance to represent
# a student's academic engagement.

df["Productivity_Index"] = df["Study_Hours"] * df["Attendance"]

df



# ==========================================================
# CELL 4 – FEATURE TRANSFORMATION
# ==========================================================
# Sometimes data can be skewed (not evenly distributed).
#
# Log transformation is commonly used to reduce skewness
# and make data more suitable for machine learning models.

df["Log_Study_Hours"] = np.log(df["Study_Hours"])

df



# ==========================================================
# CELL 5 – BINNING (CREATING CATEGORIES FROM NUMBERS)
# ==========================================================
# We can convert continuous numerical data into categories.
#
# Example:
# Attendance levels:
# Low
# Medium
# High

df["Attendance_Level"] = pd.cut(
    df["Attendance"],
    bins=[0, 75, 90, 100],
    labels=["Low", "Medium", "High"]
)

df



# ==========================================================
# CELL 6 – FEATURE SELECTION
# ==========================================================
# Sometimes not all features are useful.
# Removing unnecessary features can simplify the model
# and improve performance.

selected_features = df[["Study_Hours", "Attendance", "Productivity_Index"]]

selected_features



# ==========================================================
# CELL 7 – SUMMARY
# ==========================================================
# Feature Engineering helps machine learning models by:
#
# 1. Creating meaningful features
# 2. Transforming data into better representations
# 3. Removing irrelevant features
#
# In real-world ML projects, feature engineering is often
# an iterative and creative process.
#
# Skilled data scientists spend significant time designing
# better features before training models.
```
