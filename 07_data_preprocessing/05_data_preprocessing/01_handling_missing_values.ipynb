```python
# ==========================================================
# DATA PREPROCESSING – HANDLING MISSING VALUES
# ==========================================================
# In real-world datasets, missing values are extremely common.
# These missing values may occur due to:
#
# • Data entry errors
# • Sensors failing to record values
# • Users skipping information
# • Data merging from multiple sources
#
# Machine Learning models cannot handle missing values directly.
# So before training a model we must clean the dataset.
#
# Common techniques for handling missing values:
#
# 1. Removing rows or columns that contain missing values
# 2. Filling missing values with statistical measures
#       - Mean
#       - Median
#       - Mode
# 3. Forward fill or backward fill
# 4. Using ML models to predict missing values
#
# In this notebook we will learn the basic and most common methods.
# ==========================================================



# ==========================================================
# CELL 1 - IMPORT LIBRARIES
# ==========================================================
# pandas → used for data manipulation and dataset handling
# numpy → used for numerical operations and creating NaN values

import pandas as pd
import numpy as np


# ===============================
# CELL 2 – CREATE A SAMPLE DATASET WITH MISSING VALUES
# ==========================================================
# Here we create a small dataset that contains missing values
# so we can practice different preprocessing techniques.

data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Age": [23, np.nan, 35, 29, np.nan],
    "Salary": [50000, 60000, np.nan, 58000, 62000],
    "City": ["Delhi", "Mumbai", "Delhi", np.nan, "Chennai"]
}

df = pd.DataFrame(data)

df

# ==========================================================
# CELL3 - CHECK FOR MISSING VALUES
# ==========================================================
# isnull() returns True wherever a value is missing (NaN).
# This helps us quickly identify which cells contain missing data.

df.isnull()


# ==========================================================
# CELL 5 – Count Missing Values
# ==========================================================
# isnull() returns True wherever a value is missing (NaN).
# sum() counts how many missing values exist in each column.



df.isnull().sum()


# ===============================
# CELL 5 – DROP MISSING VALUES
# ===============================

df_drop = df.dropna()

df_drop


# ===============================
# CELL 6 – FILL MISSING VALUES (MEAN)
# ===============================

df_fill_mean = df.copy()

df_fill_mean["Age"] = df_fill_mean["Age"].fillna(df["Age"].mean())

df_fill_mean


# ===============================
# CELL 7 – FILL MISSING VALUES (MEDIAN)
# ===============================

df_fill_median = df.copy()

df_fill_median["Salary"] = df_fill_median["Salary"].fillna(df["Salary"].median())

df_fill_median


# ===============================
# CELL 8 – FILL CATEGORICAL VALUES
# ===============================

df_fill_city = df.copy()

df_fill_city["City"] = df_fill_city["City"].fillna("Unknown")

df_fill_city


# ===============================
# CELL 9 – FORWARD FILL
# ===============================

df_ffill = df.fillna(method="ffill")

df_ffill


# ===============================
# CELL 10 – BACKWARD FILL
# ===============================

df_bfill = df.fillna(method="bfill")

df_bfill
```
