```python
# ==========================================================
# DATA PREPROCESSING – FEATURE SCALING
# ==========================================================
# Feature Scaling is a very important step in machine learning.
#
# In many datasets, features can have very different ranges.
#
# Example:
# Age → 18 to 60
# Salary → 20000 to 100000
#
# Machine learning algorithms calculate distances between points.
# If one feature has a much larger range, it can dominate the model
# and affect the results.
#
# For example in KNN or Gradient Descent based models:
# A large value feature like salary may overpower a smaller
# feature like age.
#
# To solve this, we scale the features so that they are in a
# similar range.
#
# Two common techniques are:
#
# 1. NORMALIZATION (Min-Max Scaling)
#    Values are scaled between 0 and 1
#
#    Formula:
#    X_scaled = (X - X_min) / (X_max - X_min)
#
#
# 2. STANDARDIZATION (Z-Score Scaling)
#    Data is transformed so that it has:
#    Mean = 0
#    Standard Deviation = 1
#
#    Formula:
#    X_scaled = (X - Mean) / Standard Deviation
#
# Standardization is commonly used in many ML models like:
# Logistic Regression
# Support Vector Machines
# Neural Networks
#
# In this notebook we will learn both methods.
# ==========================================================



# ==========================================================
# CELL 1 – IMPORT LIBRARIES
# ==========================================================
# pandas → dataset manipulation
# sklearn.preprocessing → scaling tools

import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler



# ==========================================================
# CELL 2 – CREATE SAMPLE DATASET
# ==========================================================
# Here we create a dataset where features have very
# different ranges.

data = {
    "Age": [22, 25, 47, 52, 46],
    "Salary": [20000, 35000, 60000, 80000, 75000]
}

df = pd.DataFrame(data)

df



# ==========================================================
# CELL 3 – UNDERSTAND THE SCALE DIFFERENCE
# ==========================================================
# Age ranges around 20–50
# Salary ranges around 20000–80000
#
# This difference in magnitude can affect many ML algorithms.

df.describe()



# ==========================================================
# CELL 4 – NORMALIZATION (MIN-MAX SCALING)
# ==========================================================
# MinMaxScaler converts values into a range between 0 and 1.
#
# Formula:
# (X - X_min) / (X_max - X_min)

minmax_scaler = MinMaxScaler()

normalized_data = minmax_scaler.fit_transform(df)

normalized_df = pd.DataFrame(normalized_data, columns=df.columns)

normalized_df



# ==========================================================
# CELL 5 – STANDARDIZATION (Z-SCORE SCALING)
# ==========================================================
# StandardScaler transforms the data so that:
#
# Mean = 0
# Standard deviation = 1
#
# This method is widely used in ML algorithms.

standard_scaler = StandardScaler()

standardized_data = standard_scaler.fit_transform(df)

standardized_df = pd.DataFrame(standardized_data, columns=df.columns)

standardized_df



# ==========================================================
# CELL 6 – COMPARISON
# ==========================================================
# Original data vs Normalized vs Standardized

print("Original Data\n")
print(df)

print("\nNormalized Data\n")
print(normalized_df)

print("\nStandardized Data\n")
print(standardized_df)



# ==========================================================
# CELL 7 – WHEN TO USE WHICH?
# ==========================================================
# Normalization is often used when:
# - Data has a fixed range
# - Algorithms like KNN, Neural Networks
#
# Standardization is often used when:
# - Data distribution is unknown
# - Algorithms like Logistic Regression, SVM
#
# In practice, StandardScaler is used very frequently.
```
