```python
# =========================================
# CELL 1 - TITLE AND INTRODUCTION (MARKDOWN)
# =========================================
"""
MATPLOTLIB BASICS

This notebook introduces basic data visualization using Matplotlib.

Matplotlib is widely used for:
- Line plots
- Scatter plots
- Bar charts
- Histograms
- Pie charts
- Subplots

These visualizations are important for Exploratory Data Analysis (EDA)
in Machine Learning and Data Science.
"""


# =========================================
# CELL 2 - IMPORT LIBRARIES
# =========================================
import matplotlib.pyplot as plt
import numpy as np


# =========================================
# CELL 3 - SIMPLE LINE PLOT
# =========================================
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y)

plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.show()


# =========================================
# CELL 4 - LINE PLOT WITH MARKERS
# =========================================
x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x,y, marker='o', color='green')

plt.title("Line Plot with Markers")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()


# =========================================
# CELL 5 - SCATTER PLOT
# =========================================
x = [5,7,8,7,2,17,2,9]
y = [99,86,87,88,100,86,103,87]

plt.scatter(x,y)

plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()


# =========================================
# CELL 6 - BAR CHART
# =========================================
categories = ["A","B","C","D"]
values = [5,7,3,4]

plt.bar(categories, values)

plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.show()


# =========================================
# CELL 7 - HISTOGRAM
# =========================================
data = np.random.randn(1000)

plt.hist(data, bins=30)

plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()


# =========================================
# CELL 8 - PIE CHART
# =========================================
labels = ['Python','Java','C++','JavaScript']
sizes = [40,25,20,15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.title("Programming Language Popularity")

plt.show()


# =========================================
# CELL 9 - MULTIPLE PLOTS (SUBPLOTS)
# =========================================
x = [1,2,3,4]
y1 = [10,20,30,40]
y2 = [40,30,20,10]

plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title("Plot 2")

plt.show()


# =========================================
# CELL 10 - LEGEND AND GRID
# =========================================
x = [1,2,3,4]

y1 = [10,20,30,40]
y2 = [40,30,20,10]

plt.plot(x,y1, label="Growth")
plt.plot(x,y2, label="Decline")

plt.title("Plot with Legend")
plt.xlabel("X")
plt.ylabel("Y")

plt.legend()
plt.grid()

plt.show()


# =========================================
# CELL 11 - PRACTICE EXERCISE
# =========================================
"""
PRACTICE EXERCISES

1. Create a line plot showing temperature changes over 7 days
2. Create a bar chart for student marks in 5 subjects
3. Generate 500 random numbers and plot a histogram
4. Create subplots for:
   - sine wave
   - cosine wave
"""


# =========================================
# CELL 12 - SINE WAVE EXAMPLE
# =========================================
x = np.linspace(0,10,100)

y = np.sin(x)

plt.plot(x,y)

plt.title("Sine Wave")

plt.show()
```
