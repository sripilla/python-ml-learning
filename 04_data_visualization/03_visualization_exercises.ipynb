```python
# =========================================
# CELL 1 - TITLE AND INTRODUCTION
# =========================================
"""
DATA VISUALIZATION PRACTICE EXERCISES

This notebook contains practice exercises using
Matplotlib and Seaborn.

Goals:
- Strengthen plotting skills
- Perform basic exploratory data analysis
- Understand feature relationships
"""


# =========================================
# CELL 2 - IMPORT LIBRARIES
# =========================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# =========================================
# CELL 3 - LOAD SAMPLE DATASET
# =========================================
tips = sns.load_dataset("tips")

tips.head()


# =========================================
# CELL 4 - DATASET OVERVIEW
# =========================================
print("Dataset Shape:", tips.shape)
print("\nColumns:\n", tips.columns)
print("\nSummary:\n")
print(tips.describe())


# =========================================
# CELL 5 - HISTOGRAM OF TOTAL BILL
# =========================================
plt.hist(tips["total_bill"], bins=20)

plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")

plt.show()


# =========================================
# CELL 6 - SCATTER PLOT (BILL VS TIP)
# =========================================
plt.scatter(tips["total_bill"], tips["tip"])

plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()


# =========================================
# CELL 7 - BAR PLOT USING SEABORN
# =========================================
sns.barplot(x="day", y="total_bill", data=tips)

plt.title("Average Total Bill per Day")

plt.show()


# =========================================
# CELL 8 - BOX PLOT (OUTLIER DETECTION)
# =========================================
sns.boxplot(x="day", y="total_bill", data=tips)

plt.title("Box Plot of Total Bill by Day")

plt.show()


# =========================================
# CELL 9 - CORRELATION HEATMAP
# =========================================
corr = tips.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()


# =========================================
# CELL 10 - PAIRPLOT (RELATIONSHIPS)
# =========================================
sns.pairplot(tips)

plt.show()


# =========================================
# CELL 11 - PRACTICE TASKS
# =========================================
"""
TRY THESE EXERCISES

1. Plot histogram of tip column
2. Plot scatter of size vs total_bill
3. Plot boxplot of smoker vs total_bill
4. Create a heatmap of correlations
"""


# =========================================
# CELL 12 - OPTIONAL PRACTICE WITH RANDOM DATA
# =========================================
data = np.random.randn(500)

plt.hist(data, bins=25)

plt.title("Random Data Distribution")

plt.show()
```
