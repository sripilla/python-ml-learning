```python
# =========================================
# CELL 1 - TITLE AND INTRODUCTION
# =========================================
"""
SEABORN STATISTICAL VISUALIZATION

Seaborn is built on top of Matplotlib and is used for
statistical data visualization.

It helps visualize:
- Distributions
- Relationships between variables
- Correlations
- Outliers
- Dataset patterns

This library is heavily used in:
Machine Learning
Exploratory Data Analysis (EDA)
Data Science Projects
"""


# =========================================
# CELL 2 - IMPORT LIBRARIES
# =========================================
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# =========================================
# CELL 3 - LOAD SAMPLE DATASET
# =========================================
tips = sns.load_dataset("tips")

tips.head()


# =========================================
# CELL 4 - DATASET INFORMATION
# =========================================
tips.info()


# =========================================
# CELL 5 - HISTOGRAM / DISTRIBUTION PLOT
# =========================================
sns.histplot(tips["total_bill"])

plt.title("Distribution of Total Bill")

plt.show()


# =========================================
# CELL 6 - KDE DISTRIBUTION
# =========================================
sns.kdeplot(tips["total_bill"], fill=True)

plt.title("KDE Distribution")

plt.show()


# =========================================
# CELL 7 - SCATTER PLOT
# =========================================
sns.scatterplot(x="total_bill", y="tip", data=tips)

plt.title("Total Bill vs Tip")

plt.show()


# =========================================
# CELL 8 - BOX PLOT
# =========================================
sns.boxplot(x="day", y="total_bill", data=tips)

plt.title("Box Plot - Day vs Total Bill")

plt.show()


# =========================================
# CELL 9 - VIOLIN PLOT
# =========================================
sns.violinplot(x="day", y="total_bill", data=tips)

plt.title("Violin Plot")

plt.show()


# =========================================
# CELL 10 - PAIR PLOT
# =========================================
sns.pairplot(tips)

plt.show()


# =========================================
# CELL 11 - CORRELATION HEATMAP
# =========================================
corr = tips.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()


# =========================================
# CELL 12 - PRACTICE EXERCISES
# =========================================
"""
PRACTICE EXERCISES

1. Create a scatter plot of total_bill vs size
2. Create a boxplot for smoker vs total_bill
3. Create a heatmap for correlation matrix
4. Use pairplot to visualize relationships
"""
```
