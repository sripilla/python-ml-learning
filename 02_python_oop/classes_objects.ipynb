```python
"""
1. What is Object Oriented Programming (OOP)
2. What is a Class
3. Creating a Class
4. Creating an Object
5. Class Attributes
6. Methods in a Class
7. The __init__ Constructor
8. Creating Multiple Objects
9. Practice Exercises
"""

# What is Object Oriented Programming (OOP)
# OOP is a programming paradigm based on objects and classes.
# A class is a blueprint, and objects are instances of that class.


# Creating a Class
# A class is defined using the keyword 'class'.

class Person:
    name = "Unknown"
    age = 0


# Creating an Object
# An object is created from a class.

person1 = Person()

print(person1.name)
print(person1.age)


# Modifying Object Attributes

person1.name = "Likitha"
person1.age = 20

print(person1.name)
print(person1.age)


# Class Attributes
# Attributes are variables that belong to a class.

class Student:
    college = "MIT Manipal"

student1 = Student()

print(student1.college)


# Methods in a Class
# Methods are functions defined inside a class.

class Greeting:

    def say_hello(self):
        print("Hello from the class")

g = Greeting()

g.say_hello()


# The __init__ Constructor
# The constructor runs automatically when an object is created.

class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

student1 = Student("Likitha", 20)

print(student1.name)
print(student1.age)


# Creating Multiple Objects

student1 = Student("Likitha", 20)
student2 = Student("Rahul", 22)

print(student1.name)
print(student2.name)


# Practice Exercise 1
# Create a class called Car with attributes brand and model.

class Car:
    brand = "Toyota"
    model = "Corolla"

car1 = Car()

print(car1.brand)
print(car1.model)


# Practice Exercise 2
# Create a class with a method that prints a message.

class Welcome:

    def message(self):
        print("Welcome to Python OOP")

obj = Welcome()

obj.message()


# Practice Exercise 3
# Create a class Person with name and age using constructor.

class Person:

    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("John", 30)

print(p1.name)
print(p1.age)
```
