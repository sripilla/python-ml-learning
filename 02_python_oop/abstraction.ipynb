```python
"""
1. What is Abstraction
2. Abstract Classes
3. Abstract Methods
4. Implementing Abstract Classes
5. Real World Example
6. Practice Exercises
"""

# What is Abstraction
# Abstraction means hiding implementation details
# and showing only the essential features to the user.
# In Python, abstraction is implemented using abstract classes.


# Importing the required module
# Python provides the abc module for abstract classes.

from abc import ABC, abstractmethod


# Abstract Class
# A class that contains one or more abstract methods.

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass


# Implementing Abstract Methods
# Any class inheriting from Animal must implement sound().

class Dog(Animal):

    def sound(self):
        print("Dog barks")


class Cat(Animal):

    def sound(self):
        print("Cat meows")


d = Dog()
c = Cat()

d.sound()
c.sound()


# Another Example

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass


class Rectangle(Shape):

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


r = Rectangle(5,3)

print(r.area())


# Practice Exercise 1
# Create an abstract class Vehicle with method start().

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass


class Car(Vehicle):

    def start(self):
        print("Car starts with key")


car = Car()

car.start()


# Practice Exercise 2
# Create an abstract class Shape with method area().

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass


class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius


c = Circle(5)

print(c.area())
```
