```python
"""
1. What is Encapsulation
2. Public Variables
3. Protected Variables
4. Private Variables
5. Accessing Private Variables
6. Getter Methods
7. Setter Methods
8. Practice Exercises
"""

# What is Encapsulation
# Encapsulation means restricting direct access to certain variables
# and methods inside a class.
# It helps protect data from accidental modification.


# Public Variables
# Public variables can be accessed from anywhere.

class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age


s = Student("Likitha", 20)

print(s.name)
print(s.age)


# Protected Variables
# Protected variables are indicated by a single underscore (_).
# They should not be accessed directly outside the class,
# but Python does not strictly prevent access.

class Student:

    def __init__(self, name):
        self._name = name


s = Student("Likitha")

print(s._name)


# Private Variables
# Private variables use double underscore (__).
# They cannot be accessed directly outside the class.

class Student:

    def __init__(self, name):
        self.__name = name

    def display(self):
        print(self.__name)


s = Student("Likitha")

s.display()


# Accessing Private Variables
# Direct access will cause an error.

class Person:

    def __init__(self, age):
        self.__age = age

    def show_age(self):
        print(self.__age)


p = Person(30)

p.show_age()


# Getter Methods
# Getter methods allow controlled access to private variables.

class Employee:

    def __init__(self, salary):
        self.__salary = salary

    def get_salary(self):
        return self.__salary


emp = Employee(50000)

print(emp.get_salary())


# Setter Methods
# Setter methods allow updating private variables safely.

class Employee:

    def __init__(self, salary):
        self.__salary = salary

    def set_salary(self, salary):
        self.__salary = salary

    def get_salary(self):
        return self.__salary


emp = Employee(50000)

emp.set_salary(60000)

print(emp.get_salary())


# Practice Exercise 1
# Create a class BankAccount with private balance.

class BankAccount:

    def __init__(self, balance):
        self.__balance = balance

    def get_balance(self):
        return self.__balance


account = BankAccount(1000)

print(account.get_balance())


# Practice Exercise 2
# Update the balance using a setter method.

class BankAccount:

    def __init__(self, balance):
        self.__balance = balance

    def set_balance(self, balance):
        self.__balance = balance

    def get_balance(self):
        return self.__balance


account = BankAccount(1000)

account.set_balance(1500)

print(account.get_balance())


# Practice Exercise 3
# Create a class Person with private age and getter method.

class Person:

    def __init__(self, age):
        self.__age = age

    def get_age(self):
        return self.__age


p = Person(25)

print(p.get_age())
```
