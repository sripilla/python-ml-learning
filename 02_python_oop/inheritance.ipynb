```python
"""
1. What is Inheritance
2. Creating a Parent Class
3. Creating a Child Class
4. Inheriting Methods
5. Overriding Methods
6. Using super()
7. Practice Exercises
"""

# What is Inheritance
# Inheritance allows a class (child class) to inherit properties and methods
# from another class (parent class).
# This helps reuse code and avoid repetition.


# Creating a Parent Class

class Animal:

    def speak(self):
        print("Animals make sounds")


# Creating a Child Class
# The child class inherits from the parent class.

class Dog(Animal):
    pass


dog = Dog()

dog.speak()


# Another Example of Inheritance

class Vehicle:

    def start(self):
        print("Vehicle started")


class Car(Vehicle):
    pass


car = Car()

car.start()


# Overriding Methods
# A child class can override methods from the parent class.

class Animal:

    def speak(self):
        print("Animal makes a sound")


class Dog(Animal):

    def speak(self):
        print("Dog barks")


dog = Dog()

dog.speak()


# Using super()
# super() allows calling a method from the parent class.

class Person:

    def __init__(self, name):
        self.name = name


class Student(Person):

    def __init__(self, name, course):
        super().__init__(name)
        self.course = course


student = Student("Likitha", "AI/ML")

print(student.name)
print(student.course)


# Practice Exercise 1
# Create a parent class called Bird with a method fly().

class Bird:

    def fly(self):
        print("Birds can fly")


# Create a child class Eagle inheriting from Bird.

class Eagle(Bird):
    pass


eagle = Eagle()

eagle.fly()


# Practice Exercise 2
# Override a method in the child class.

class Bird:

    def sound(self):
        print("Bird makes a sound")


class Parrot(Bird):

    def sound(self):
        print("Parrot talks")


p = Parrot()

p.sound()


# Practice Exercise 3
# Use super() in a child class.

class Person:

    def __init__(self, name):
        self.name = name


class Teacher(Person):

    def __init__(self, name, subject):
        super().__init__(name)
        self.subject = subject


t = Teacher("John", "Math")

print(t.name)
print(t.subject)
```
