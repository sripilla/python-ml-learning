```python
"""
1. What is Polymorphism
2. Function Polymorphism
3. Operator Polymorphism
4. Method Overriding
5. Polymorphism with Multiple Classes
6. Practice Exercises
"""

# What is Polymorphism
# Polymorphism means "many forms".
# The same function or method name can behave differently depending on the object.


# Function Polymorphism
# Built-in functions can work with different data types.

print(len("Python"))
print(len([1, 2, 3, 4]))
print(len((10, 20, 30)))


# Operator Polymorphism
# The + operator works differently depending on the data type.

print(10 + 5)              # Addition
print("Hello " + "World")  # String concatenation
print([1,2] + [3,4])       # List merging


# Method Overriding
# A child class can provide a different implementation of a method
# defined in the parent class.

class Animal:

    def sound(self):
        print("Animal makes a sound")


class Dog(Animal):

    def sound(self):
        print("Dog barks")


class Cat(Animal):

    def sound(self):
        print("Cat meows")


dog = Dog()
cat = Cat()

dog.sound()
cat.sound()


# Polymorphism with Multiple Classes
# Different classes implementing the same method name.

class Bird:

    def fly(self):
        print("Bird can fly")


class Airplane:

    def fly(self):
        print("Airplane flies in the sky")


objects = [Bird(), Airplane()]

for obj in objects:
    obj.fly()


# Practice Exercise 1
# Create two classes with the same method name.

class Car:

    def move(self):
        print("Car drives on road")


class Boat:

    def move(self):
        print("Boat sails on water")


c = Car()
b = Boat()

c.move()
b.move()


# Practice Exercise 2
# Demonstrate operator polymorphism.

print(5 + 10)
print("Data " + "Science")


# Practice Exercise 3
# Use a loop to call the same method on different objects.

class Dog:

    def sound(self):
        print("Dog barks")


class Cat:

    def sound(self):
        print("Cat meows")


animals = [Dog(), Cat()]

for animal in animals:
    animal.sound()
```
