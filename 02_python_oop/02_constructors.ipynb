```python
"""
1. What is a Constructor
2. Default Constructor
3. Parameterized Constructor
4. Using Constructor to Initialize Attributes
5. Creating Multiple Objects
6. Constructor with Default Values
7. Practice Exercises
"""

# What is a Constructor
# A constructor is a special method that is automatically called
# when an object of a class is created.
# In Python the constructor method is __init__().


# Default Constructor
# A constructor that does not take parameters.

class Person:

    def __init__(self):
        print("Constructor called")

p = Person()


# Parameterized Constructor
# A constructor that takes parameters.

class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age


s = Student("Likitha", 20)

print(s.name)
print(s.age)


# Using Constructor to Initialize Attributes

class Car:

    def __init__(self, brand, model):
        self.brand = brand
        self.model = model


car1 = Car("Toyota", "Corolla")

print(car1.brand)
print(car1.model)


# Creating Multiple Objects

car1 = Car("Toyota", "Corolla")
car2 = Car("Honda", "Civic")

print(car1.brand)
print(car2.brand)


# Constructor with Default Values

class Employee:

    def __init__(self, name, salary=30000):
        self.name = name
        self.salary = salary


emp1 = Employee("John")
emp2 = Employee("Alice", 50000)

print(emp1.salary)
print(emp2.salary)


# Practice Exercise 1
# Create a class Book with title and author.

class Book:

    def __init__(self, title, author):
        self.title = title
        self.author = author


book1 = Book("Python Basics", "John")

print(book1.title)
print(book1.author)


# Practice Exercise 2
# Create two objects of a class Person.

class Person:

    def __init__(self, name):
        self.name = name


p1 = Person("Alice")
p2 = Person("Bob")

print(p1.name)
print(p2.name)


# Practice Exercise 3
# Create a class Rectangle and calculate area.

class Rectangle:

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


r = Rectangle(5, 3)

print(r.area())
```
