```python
"""
1. Detecting Missing Values
2. Dropping Missing Values
3. Filling Missing Values
4. Checking Duplicates
5. Removing Duplicates
6. Renaming Columns
7. Practice Exercises
"""

# Import Pandas

import pandas as pd
import numpy as np


# Creating a sample dataset with missing values

data = {
    "Name":["Alice","Bob","Charlie","David","Emma"],
    "Age":[25,np.nan,35,28,np.nan],
    "Score":[85,90,np.nan,92,88]
}

df = pd.DataFrame(data)

print(df)


# Detecting Missing Values
# isnull() identifies missing values.

print(df.isnull())


# Count missing values in each column

print(df.isnull().sum())


# Dropping Missing Values
# dropna() removes rows containing missing values.

clean_df = df.dropna()

print(clean_df)


# Filling Missing Values
# fillna() replaces missing values.

df["Age"] = df["Age"].fillna(df["Age"].mean())

print(df)


# Filling with a constant value

df["Score"] = df["Score"].fillna(0)

print(df)


# Checking Duplicates
# duplicated() identifies duplicate rows.

print(df.duplicated())


# Removing Duplicates

df = df.drop_duplicates()

print(df)


# Renaming Columns

df = df.rename(columns={"Score":"Exam_Score"})

print(df)


# Practice Exercise 1
# Check missing values.

print(df.isnull().sum())


# Practice Exercise 2
# Fill missing values in Age with the mean.

df["Age"] = df["Age"].fillna(df["Age"].mean())

print(df)


# Practice Exercise 3
# Rename column Age to Student_Age.

df = df.rename(columns={"Age":"Student_Age"})

print(df)
```
