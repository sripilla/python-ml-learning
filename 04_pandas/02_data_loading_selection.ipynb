```python
"""
1. Loading Data from CSV
2. Viewing Dataset
3. Selecting Columns
4. Selecting Multiple Columns
5. Selecting Rows using loc
6. Selecting Rows using iloc
7. Conditional Filtering
8. Practice Exercises
"""

# Import Pandas

import pandas as pd


# Loading Data from CSV
# Pandas can load datasets stored in CSV files.

# Example:
# df = pd.read_csv("data.csv")

# For demonstration we will create a sample DataFrame.

data = {
    "Name":["Alice","Bob","Charlie","David"],
    "Age":[25,30,35,28],
    "Score":[85,90,88,92]
}

df = pd.DataFrame(data)

print(df)


# Viewing Dataset

print(df.head())
print(df.tail())


# Selecting Columns

print(df["Name"])


# Selecting Multiple Columns

print(df[["Name","Score"]])


# Selecting Rows using loc
# loc selects rows by label.

print(df.loc[0])
print(df.loc[1])


# Selecting Rows using iloc
# iloc selects rows by integer position.

print(df.iloc[0])
print(df.iloc[1])


# Selecting specific row and column

print(df.loc[0,"Name"])
print(df.iloc[2,1])


# Conditional Filtering
# Select rows based on conditions.

print(df[df["Age"] > 28])


# Multiple condition filtering

print(df[(df["Age"] > 25) & (df["Score"] > 88)])


# Practice Exercise 1
# Display the Age column.

print(df["Age"])


# Practice Exercise 2
# Display rows where Score is greater than 88.

print(df[df["Score"] > 88])


# Practice Exercise 3
# Select Name and Age columns.

print(df[["Name","Age"]])
```
