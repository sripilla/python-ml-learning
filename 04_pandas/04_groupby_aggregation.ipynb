```python
"""
1. What is GroupBy
2. Grouping Data
3. Aggregation Functions
4. Multiple Aggregations
5. Grouping by Multiple Columns
6. Sorting Grouped Data
7. Practice Exercises
"""

# Import Pandas

import pandas as pd


# Creating a sample dataset

data = {
    "Department":["IT","HR","IT","HR","Finance","Finance"],
    "Employee":["Alice","Bob","Charlie","David","Emma","Frank"],
    "Salary":[60000,50000,65000,52000,70000,72000]
}

df = pd.DataFrame(data)

print(df)


# What is GroupBy
# groupby() groups rows that have the same value in a column.

grouped = df.groupby("Department")

print(grouped)


# Grouping Data
# Calculate average salary by department.

print(df.groupby("Department")["Salary"].mean())


# Aggregation Functions
# Other aggregation functions include sum, max, min, count.

print(df.groupby("Department")["Salary"].sum())
print(df.groupby("Department")["Salary"].max())
print(df.groupby("Department")["Salary"].count())


# Multiple Aggregations

print(df.groupby("Department")["Salary"].agg(["mean","max","min"]))


# Grouping by Multiple Columns

data = {
    "Department":["IT","IT","HR","HR"],
    "Gender":["M","F","M","F"],
    "Salary":[60000,65000,50000,52000]
}

df = pd.DataFrame(data)

print(df.groupby(["Department","Gender"])["Salary"].mean())


# Sorting Grouped Data

print(df.groupby("Department")["Salary"].mean().sort_values())


# Practice Exercise 1
# Calculate average salary by department.

data = {
    "Department":["Sales","Sales","IT","IT"],
    "Salary":[40000,42000,60000,65000]
}

df = pd.DataFrame(data)

print(df.groupby("Department")["Salary"].mean())


# Practice Exercise 2
# Find maximum salary by department.

print(df.groupby("Department")["Salary"].max())


# Practice Exercise 3
# Count employees in each department.

print(df.groupby("Department")["Salary"].count())
```
