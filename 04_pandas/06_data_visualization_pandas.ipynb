```python
"""
1. Introduction to Pandas Visualization
2. Line Plot
3. Bar Chart
4. Histogram
5. Scatter Plot
6. Box Plot
7. Practice Exercises
"""

# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt


# Creating Sample Dataset

data = {
    "Year":[2018,2019,2020,2021,2022],
    "Sales":[200,250,300,280,320],
    "Profit":[50,60,70,65,80]
}

df = pd.DataFrame(data)

print(df)


# Line Plot
# Useful for showing trends over time.

df.plot(x="Year", y="Sales", kind="line", title="Sales Over Years")

plt.show()


# Bar Chart
# Useful for comparing categories.

df.plot(x="Year", y="Profit", kind="bar", title="Profit by Year")

plt.show()


# Histogram
# Shows distribution of values.

df["Sales"].plot(kind="hist", bins=5, title="Sales Distribution")

plt.show()


# Scatter Plot
# Shows relationship between two variables.

df.plot(x="Sales", y="Profit", kind="scatter", title="Sales vs Profit")

plt.show()


# Box Plot
# Shows spread and outliers in data.

df[["Sales","Profit"]].plot(kind="box", title="Box Plot")

plt.show()


# Practice Exercise 1
# Create a line plot for Profit.

df.plot(x="Year", y="Profit", kind="line", title="Profit Trend")

plt.show()


# Practice Exercise 2
# Create a histogram for Profit.

df["Profit"].plot(kind="hist", bins=5)

plt.show()


# Practice Exercise 3
# Create a scatter plot between Sales and Profit.

df.plot(x="Sales", y="Profit", kind="scatter")

plt.show()
```
