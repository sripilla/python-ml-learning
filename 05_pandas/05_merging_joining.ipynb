```python
"""
1. What is Merging
2. Creating Sample DataFrames
3. Inner Merge
4. Left Merge
5. Right Merge
6. Outer Merge
7. Joining DataFrames
8. Concatenating DataFrames
9. Practice Exercises
"""

# Import Pandas

import pandas as pd


# Creating Sample DataFrames

students = pd.DataFrame({
    "StudentID":[1,2,3],
    "Name":["Alice","Bob","Charlie"]
})

marks = pd.DataFrame({
    "StudentID":[1,2,4],
    "Marks":[85,90,75]
})

print(students)
print(marks)


# Inner Merge
# Keeps only rows where the key exists in both DataFrames.

merged = pd.merge(students, marks, on="StudentID", how="inner")

print(merged)


# Left Merge
# Keeps all rows from the left DataFrame.

left_merge = pd.merge(students, marks, on="StudentID", how="left")

print(left_merge)


# Right Merge
# Keeps all rows from the right DataFrame.

right_merge = pd.merge(students, marks, on="StudentID", how="right")

print(right_merge)


# Outer Merge
# Keeps all rows from both DataFrames.

outer_merge = pd.merge(students, marks, on="StudentID", how="outer")

print(outer_merge)


# Joining DataFrames
# join() combines DataFrames based on index.

df1 = pd.DataFrame({
    "A":[1,2,3]
}, index=["a","b","c"])

df2 = pd.DataFrame({
    "B":[4,5,6]
}, index=["a","b","c"])

joined = df1.join(df2)

print(joined)


# Concatenating DataFrames
# concat() stacks DataFrames vertically or horizontally.

df1 = pd.DataFrame({"A":[1,2]})
df2 = pd.DataFrame({"A":[3,4]})

result = pd.concat([df1,df2])

print(result)


# Practice Exercise 1
# Merge two DataFrames using inner join.

df1 = pd.DataFrame({
    "ID":[1,2,3],
    "Name":["Tom","Jerry","Spike"]
})

df2 = pd.DataFrame({
    "ID":[1,2,4],
    "Score":[80,85,90]
})

print(pd.merge(df1,df2,on="ID",how="inner"))


# Practice Exercise 2
# Perform left join.

print(pd.merge(df1,df2,on="ID",how="left"))


# Practice Exercise 3
# Concatenate two DataFrames.

df3 = pd.DataFrame({"A":[10,20]})
df4 = pd.DataFrame({"A":[30,40]})

print(pd.concat([df3,df4]))
```
