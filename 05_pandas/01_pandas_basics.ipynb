```python
"""
1. What is Pandas
2. Installing and Importing Pandas
3. Pandas Series
4. Creating DataFrames
5. DataFrame from Dictionary
6. Viewing Data
7. Practice Exercises
"""

# What is Pandas
# Pandas is a Python library used for data analysis and data manipulation.
# It provides powerful data structures like Series and DataFrames.


# Importing Pandas

import pandas as pd


# Pandas Series
# A Series is a one-dimensional labeled array.

data = [10,20,30,40]

series = pd.Series(data)

print(series)


# Series with custom index

series = pd.Series([100,200,300], index=["a","b","c"])

print(series)


# Creating DataFrames
# A DataFrame is a two-dimensional table-like structure.

data = {
    "Name":["Alice","Bob","Charlie"],
    "Age":[25,30,35],
    "City":["Delhi","Mumbai","Bangalore"]
}

df = pd.DataFrame(data)

print(df)


# DataFrame from Dictionary

student_data = {
    "Name":["Likitha","Rahul","Anita"],
    "Marks":[90,85,88]
}

df = pd.DataFrame(student_data)

print(df)


# Viewing Data

print(df.head())
print(df.tail())


# Basic Information

print(df.info())


# Practice Exercise 1
# Create a Series of numbers.

numbers = pd.Series([5,10,15,20])

print(numbers)


# Practice Exercise 2
# Create a DataFrame of students.

data = {
    "Name":["John","Emma","Mike"],
    "Score":[80,92,85]
}

df = pd.DataFrame(data)

print(df)


# Practice Exercise 3
# Display first rows of the DataFrame.

print(df.head())
```
